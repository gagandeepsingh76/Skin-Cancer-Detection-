```python
# Updated ViT training script with CBAM-like attention for tokens + 70/30 split + smoothing + full metrics
# Requirements: torch, torchvision, timm, scikit-learn, tqdm, matplotlib, seaborn, pillow
# !pip install -q torch torchvision timm scikit-learn tqdm matplotlib seaborn pillow

import os
import random
import zipfile
import time
from collections import Counter
from glob import glob

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import transforms, datasets
import timm
from tqdm import tqdm

from sklearn.metrics import (confusion_matrix, classification_report, accuracy_score,
                             roc_curve, auc, precision_recall_fscore_support)
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# ---------------- Configuration (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"            # dataset zip on Drive (if used)
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/d/train"
VAL_DIR   = "/content/dataset/d/val"

# Experiment / graph directory (keeps per-experiment plots separate)
BASE_GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
MODEL_NAME = "SkinViT224_CBAM"   # experiment name

MODEL_MAP = {
    "SkinViT224_CBAM": "vit_base_patch16_224",
    "SkinViT224":  "vit_base_patch16_224",
    "DermaViT224": "vit_base_patch16_224",
    "DermaViT384": "vit_base_patch16_384",
}

TIMM_MODEL_NAME = MODEL_MAP.get(MODEL_NAME, MODEL_NAME)

GRAPH_DIR = os.path.join(BASE_GRAPH_DIR, MODEL_NAME)
CHECKPOINT_PATH = f"/content/drive/MyDrive/{MODEL_NAME}_best.pt"
FINAL_MODEL_PATH = f"/content/drive/MyDrive/{MODEL_NAME}_final.pt"

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2
SEED = 42
PHASE1_EPOCHS = 8
PHASE2_EPOCHS = 43
N_UNFREEZE = 6
SHOW_GPU_MEM = True

SPLIT_SEED = SEED
TRAIN_FRACTION = 0.7   # 70% train, 30% val

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract (Colab) ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ---------------- Dataset combine + 70/30 split ----------------
# We combine images from TRAIN_DIR and VAL_DIR (if both exist) to create a robust 70/30 split.
def collect_image_paths(dirs):
    paths = []
    for d in dirs:
        if not d or not os.path.exists(d):
            continue
        # expect class subfolders under each dir
        for class_name in sorted(os.listdir(d)):
            class_path = os.path.join(d, class_name)
            if not os.path.isdir(class_path):
                continue
            img_pattern = os.path.join(class_path, "*")
            for p in glob(img_pattern):
                if os.path.isfile(p):
                    paths.append((p, class_name))
    return paths

combined_paths = collect_image_paths([TRAIN_DIR, VAL_DIR])
if len(combined_paths) == 0:
    raise FileNotFoundError(f"No images found in provided TRAIN_DIR/VAL_DIR paths: {TRAIN_DIR}, {VAL_DIR}")

# build class list
classes = sorted(list({c for _, c in combined_paths}))
num_classes = len(classes)
class_to_idx = {c: i for i, c in enumerate(classes)}
print(f"Detected classes ({num_classes}): {classes}")

# build list of (path, idx)
full_samples = [(p, class_to_idx[c]) for p, c in combined_paths]

# deterministic shuffle & split
random.Random(SPLIT_SEED).shuffle(full_samples)
n_total = len(full_samples)
n_train = int(TRAIN_FRACTION * n_total)
train_samples = full_samples[:n_train]
val_samples = full_samples[n_train:]
print(f"Total images: {n_total} -> train: {len(train_samples)}, val: {len(val_samples)}")

# ---------------- Transforms ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Custom Dataset from samples ----------------
from torch.utils.data import Dataset, DataLoader

class ImagePathDataset(Dataset):
    def __init__(self, samples, transform=None):
        self.samples = samples
        self.transform = transform
    def __len__(self):
        return len(self.samples)
    def __getitem__(self, idx):
        path, label = self.samples[idx]
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, label

train_dataset = ImagePathDataset(train_samples, transform=train_transform)
val_dataset   = ImagePathDataset(val_samples, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Class weights ----------------
y_train = [lbl for _, lbl in train_samples]
if num_classes > 1 and len(y_train) > 0:
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# ---------------- CBAM-like module adapted for ViT token embeddings --------------
# Input tokens shape: (B, N_tokens, C)
# Channel attention: squeeze over tokens -> (B, C) -> fc -> sigmoid -> scale channels
# Spatial/token attention: compute avg & max across channel per token -> (B, N, 1), apply conv1d (kernel 7) -> sigmoid -> scale tokens
class ChannelAttentionToken(nn.Module):
    def __init__(self, channels, reduction=8):
        super().__init__()
        self.fc1 = nn.Linear(channels, channels // reduction, bias=False)
        self.relu = nn.ReLU(inplace=True)
        self.fc2 = nn.Linear(channels // reduction, channels, bias=False)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):  # x: B x N x C
        # squeeze tokens
        x_pool = x.mean(dim=1)  # B x C
        y = self.fc1(x_pool)
        y = self.relu(y)
        y = self.fc2(y)
        y = self.sigmoid(y).unsqueeze(1)  # B x 1 x C
        return x * y  # broadcast across tokens

class SpatialAttentionToken(nn.Module):
    def __init__(self, kernel_size=7):
        super().__init__()
        padding = (kernel_size - 1) // 2
        # We'll treat token axis like spatial axis: use conv1d across tokens
        self.conv = nn.Conv1d(in_channels=2, out_channels=1, kernel_size=kernel_size, padding=padding, bias=False)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):  # x: B x N x C
        avg_out = x.mean(dim=2, keepdim=True)  # B x N x 1
        max_out, _ = x.max(dim=2, keepdim=True)  # B x N x 1
        cat = torch.cat([avg_out, max_out], dim=2)  # B x N x 2
        # need shape (B, 2, N) for conv1d
        cat = cat.permute(0, 2, 1)  # B x 2 x N
        attn = self.conv(cat)  # B x 1 x N
        attn = self.sigmoid(attn).permute(0, 2, 1)  # B x N x 1
        return x * attn  # scale tokens

class CBAMTokens(nn.Module):
    def __init__(self, channels, reduction=8, kernel_size=7):
        super().__init__()
        self.ca = ChannelAttentionToken(channels, reduction=reduction)
        self.sa = SpatialAttentionToken(kernel_size=kernel_size)
    def forward(self, x):
        x = self.ca(x)
        x = self.sa(x)
        return x

# ---------------- Build timm model and wrap with CBAMTokens ----------------
print(f"Creating timm model: {TIMM_MODEL_NAME} (experiment name: {MODEL_NAME})")
base_model = timm.create_model(TIMM_MODEL_NAME, pretrained=True, num_classes=num_classes)
base_model.to(device)
print("Base model created:", TIMM_MODEL_NAME)

# wrapper module: run through patch embedding + blocks to get tokens, apply CBAMTokens, then classifier head
class ViTWithTokenCBAM(nn.Module):
    def __init__(self, timm_model, cbam_tokens):
        super().__init__()
        self.model = timm_model
        self.cbam = cbam_tokens
        # most timm ViT models have a head called `head` or `head.fc` - we'll use model.get_classifier() / reset head
        # store original classifier for later use
        try:
            self.classifier = self.model.get_classifier()
        except Exception:
            # fallback - try `head`
            if hasattr(self.model, "head"):
                self.classifier = self.model.head
            else:
                # build a simple Linear if none exists
                embed_dim = getattr(self.model, 'embed_dim', None) or getattr(self.model, 'num_features', None)
                self.classifier = nn.Linear(embed_dim, num_classes)
        # remove the classifier inside original model to avoid double-head if forward_features returns pooled
        # but we won't call model.forward; we'll use its components.
    def forward(self, x):
        # The wrapper uses common timm ViT components when present.
        # We'll try to support: model.patch_embed, model.cls_token, model.pos_embed, model.pos_drop, model.blocks, model.norm
        if hasattr(self.model, "patch_embed") and hasattr(self.model, "pos_embed"):
            # typical ViT forward_features
            x = self.model.patch_embed(x)           # B, N, C
            # cls token + pos emb
            if hasattr(self.model, "cls_token"):
                cls_token = self.model.cls_token.expand(x.shape[0], -1, -1)  # B,1,C
                x = torch.cat((cls_token, x), dim=1)  # B, N+1, C
            if hasattr(self.model, "pos_embed"):
                # pos_embed might have shape 1, N+1, C
                x = x + self.model.pos_embed
            if hasattr(self.model, "pos_drop"):
                x = self.model.pos_drop(x)
            # transformer blocks
            if hasattr(self.model, "blocks"):
                for blk in self.model.blocks:
                    x = blk(x)
            elif hasattr(self.model, "vit") and hasattr(self.model.vit, "blocks"):
                for blk in self.model.vit.blocks:
                    x = blk(x)
            # norm if present
            if hasattr(self.model, "norm"):
                x = self.model.norm(x)
            # apply CBAM tokens (works on token sequence including cls token)
            x = self.cbam(x)   # B x (N+1) x C
            # take cls token (first token) as pooled representation
            pooled = x[:, 0]  # B x C
            # pass through classifier: many timm models expect pre_logits processing
            # Some models have pre_logits method - try to be robust
            # If classifier is nn.Linear, call it directly
            out = self.classifier(pooled)
            return out
        else:
            # fallback: call original model forward and try to inject channel attention on final feature vector
            feats = self.model.forward_features(x) if hasattr(self.model, "forward_features") else None
            if feats is None:
                return self.model(x)
            # feats typically B x C (or B x dim)
            if feats.dim() == 2:
                # channel attention on vector (squeeze-excitation style)
                w = torch.sigmoid(nn.Linear(feats.size(1), feats.size(1) // 8).to(feats.device)(feats))
                feats = feats * w
                return self.classifier(feats)
            else:
                return self.model(x)

# create cbam module sized to embed_dim
embed_dim = getattr(base_model, "embed_dim", None) or getattr(base_model, "num_features", None)
if embed_dim is None:
    embed_dim = 768  # typical for base
cbam_tokens = CBAMTokens(channels=embed_dim, reduction=8, kernel_size=7)
model = ViTWithTokenCBAM(base_model, cbam_tokens)
model.to(device)

# ---------------- Phase 1: freeze encoder, train head + CBAM ----------------
for name, p in model.named_parameters():
    p.requires_grad = False

# unfreeze classifier (and cbam) parameters only
unfroze = 0
for name, p in model.named_parameters():
    if name.startswith("classifier") or "classif" in name or "head" in name or name.startswith("cbam") or "cbam" in name or name.startswith("model.head"):
        p.requires_grad = True
        unfroze += 1
# ensure cbam parameters are trainable
for name, p in model.cbam.named_parameters():
    p.requires_grad = True
print("Phase1 trainable param groups (approx):", unfroze)

# ---------------- Loss, optimizer, scheduler ----------------
criterion = nn.CrossEntropyLoss(weight=class_weights) if class_weights is not None else nn.CrossEntropyLoss()
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=2e-4, weight_decay=1e-2)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

# ---------------- Utilities ----------------
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def smooth(x, window=5):
    if len(x) < window:
        return x
    return np.convolve(x, np.ones(window)/window, mode='same')

def compute_metrics_full(y_true, y_pred, y_prob, class_names):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    acc = accuracy_score(y_true, y_pred)
    prec_macro, rec_macro, f1_macro, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)
    prec_wt, rec_wt, f1_wt, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return {
        "accuracy": acc,
        "precision_macro": prec_macro, "recall_macro": rec_macro, "f1_macro": f1_macro,
        "precision_weighted": prec_wt, "recall_weighted": rec_wt, "f1_weighted": f1_wt,
        "report": report, "roc_auc": roc_auc, "fpr": fpr, "tpr": tpr, "macro_auc": macro_auc, "missing": missing
    }

# ---------------- Train & validate funcs ----------------
def train_one_epoch(model, loader, optimizer, device, epoch, total_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    for batch_idx, (imgs, labels) in pbar:
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        logits = model(imgs)
        loss = criterion(logits, labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 3.0)
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        b_acc = (preds == labels).float().mean().item()
        running_epoch_loss = running_loss / total
        running_epoch_acc = correct / total
        postfix = {
            "b_loss": f"{loss.item():.4f}",
            "b_acc": f"{b_acc:.3f}",
            "e_loss": f"{running_epoch_loss:.4f}",
            "e_acc": f"{running_epoch_acc:.3f}"
        }
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    return running_loss / len(loader.dataset), correct / total if total>0 else 0.0

def validate(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device)
            labels = labels.to(device)
            logits = model(imgs)
            loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset)
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ---------------- Training loop (phase 1) ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS

print("Starting Phase 1 (head-only + cbam) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    metrics = compute_metrics_full(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(metrics["accuracy"])
    history["val_auc_macro"].append(metrics["macro_auc"])
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={metrics['accuracy']:.4f} val_auc={metrics['macro_auc']:.4f}  time={time.time()-t0:.1f}s")
    # checkpoint by val loss or macro AUC
    if val_loss < best_val_loss or metrics["macro_auc"] > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, metrics["macro_auc"])
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Phase 2: unfreeze last N transformer blocks (if available) ----------------
print("Unfreezing last", N_UNFREEZE, "transformer blocks (if available).")
unfreezed_any = False
try:
    if hasattr(model.model, "blocks"):
        total_blocks = len(model.model.blocks)
        print("Transformer blocks located at model.model.blocks:", total_blocks)
        for i in range(max(0, total_blocks - N_UNFREEZE), total_blocks):
            for p in model.model.blocks[i].parameters():
                p.requires_grad = True
        unfreezed_any = True
    elif hasattr(model.model, "vit") and hasattr(model.model.vit, "blocks"):
        total_blocks = len(model.model.vit.blocks)
        print("Transformer blocks located at model.model.vit.blocks:", total_blocks)
        for i in range(max(0, total_blocks - N_UNFREEZE), total_blocks):
            for p in model.model.vit.blocks[i].parameters():
                p.requires_grad = True
        unfreezed_any = True
    if not unfreezed_any:
        print("Could not find .blocks - falling back to unfreeze entire model.")
        for p in model.parameters():
            p.requires_grad = True
except Exception as e:
    print("Error during unfreeze, falling back to full unfreeze:", e)
    for p in model.parameters():
        p.requires_grad = True

# recreate optimizer & scheduler for finetuning with lower LR
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-5, weight_decay=1e-2)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    metrics = compute_metrics_full(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(metrics["accuracy"])
    history["val_auc_macro"].append(metrics["macro_auc"])
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={metrics['accuracy']:.4f} val_auc={metrics['macro_auc']:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or metrics["macro_auc"] > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, metrics["macro_auc"])
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Final evaluation ----------------
print("Loading best model from:", CHECKPOINT_PATH)
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch="final")
metrics = compute_metrics_full(y_true_val, y_pred_val, y_prob_val, classes)

print("\nFinal Validation Accuracy: {:.4f}".format(metrics["accuracy"]))
print("\nClassification Report:\n", metrics["report"])
print("F1 macro: {:.4f}  F1 weighted: {:.4f}".format(metrics["f1_macro"], metrics["f1_weighted"]))
print("Macro AUC (mean of available class AUCs):", metrics["macro_auc"])
if metrics["missing"]:
    print("Skipped ROC for classes (no positive samples):", metrics["missing"])

# ---------------- Save & plot smoothed curves ----------------
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

# raw + smoothed accuracy
plt.figure(figsize=(9,5))
plt.plot(epochs_range, history["train_acc"], marker='o', alpha=0.6, label='Train Acc (raw)')
plt.plot(epochs_range, history["val_acc"], marker='o', alpha=0.6, label='Val Acc (raw)')
plt.plot(epochs_range, smooth(history["train_acc"], window=5), linestyle='--', label='Train Acc (smoothed)')
plt.plot(epochs_range, smooth(history["val_acc"], window=5), linestyle='--', label='Val Acc (smoothed)')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title(f"{MODEL_NAME} - Accuracy"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, f"{MODEL_NAME}_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot to:", acc_path)

# raw + smoothed loss
plt.figure(figsize=(9,5))
plt.plot(epochs_range, history["train_loss"], marker='o', alpha=0.6, label='Train Loss (raw)')
plt.plot(epochs_range, history["val_loss"], marker='o', alpha=0.6, label='Val Loss (raw)')
plt.plot(epochs_range, smooth(history["train_loss"], window=5), linestyle='--', label='Train Loss (smoothed)')
plt.plot(epochs_range, smooth(history["val_loss"], window=5), linestyle='--', label='Val Loss (smoothed)')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title(f"{MODEL_NAME} - Loss"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, f"{MODEL_NAME}_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot to:", loss_path)

# confusion matrix
cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title(f"{MODEL_NAME} - Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, f"{MODEL_NAME}_confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()
print("Saved confusion matrix to:", cm_path)

# ROC curves per class (if available)
plt.figure(figsize=(9,8))
plotted = False
for i in range(num_classes):
    if i in metrics["roc_auc"]:
        plt.plot(metrics["fpr"][i], metrics["tpr"][i], label=f"{classes[i]} (AUC={metrics['roc_auc'][i]:.3f})")
        plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title(f"{MODEL_NAME} - ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, f"{MODEL_NAME}_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot to:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

# ---------------- Save final model ----------------
torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
