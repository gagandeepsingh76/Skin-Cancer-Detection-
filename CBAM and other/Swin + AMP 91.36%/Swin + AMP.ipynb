```python
# Complete training script with Swin + AMP, MixUp/CutMix, OneCycle, WeightedSampler, etc.
# Keeps your data load paths unchanged.
# Requirements: torch, torchvision, timm, scikit-learn, tqdm, matplotlib, seaborn

import os
import random
import zipfile
import time
import math
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW, SGD
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import transforms, datasets
import timm
from tqdm import tqdm

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# ---------------- Configuration (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"            # dataset zip on Drive (if used) - KEEP
EXTRACT_PATH = "/content/dataset"                    # KEEP
TRAIN_DIR = "/content/dataset/d/train"               # KEEP
VAL_DIR   = "/content/dataset/d/val"                 # KEEP
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_swin_replacement.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_swin_replacement.pt"

MODEL_NAME = "swin_base_patch4_window7_224"   # timm Swin model
IMG_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2
SEED = 42
PHASE1_EPOCHS = 8
PHASE2_EPOCHS = 18
N_UNFREEZE = 6

# --- New technique flags & params ---
USE_AMP = True                    # mixed precision
USE_MIXUP = True                  # MixUp augmentation
MIXUP_ALPHA = 0.4
USE_CUTMIX = False                # CutMix alternative (mutually exclusive with MixUp by default)
CUTMIX_ALPHA = 1.0

USE_RANDAUG = True                # RandAugment in train transforms (lightweight)
RANDAUG_N = 2
RANDAUG_M = 9

USE_WEIGHTED_SAMPLER = True       # use WeightedRandomSampler when classes imbalanced
WEIGHTED_SAMPLER_THRESHOLD = 0.2  # if min_count / max_count < threshold -> enable sampler

USE_ONECYCLE = True               # use OneCycleLR scheduler (recommended)
MAX_LR = 2e-4
BACKBONE_LR_SCALE = 0.1           # backbone LR = MAX_LR * scale for layer-wise LR
HEAD_LR = MAX_LR                  # head lr
WEIGHT_DECAY = 1e-2

GRAD_ACCUM_STEPS = 1              # gradient accumulation
CLIP_GRAD_NORM = 1.0              # gradient clipping

USE_SWA = False                   # Stochastic Weight Averaging
SWA_START = 0.75                  # fraction of total epochs to start SWA

USE_EMA = False                   # Exponential Moving Average of weights (inference-time averaging)
EMA_DECAY = 0.999

USE_FOCAL_LOSS = False            # fallback to focal loss
FOCAL_GAMMA = 2.0
LABEL_SMOOTHING = 0.0             # set >0 for label smoothing

SHOW_GPU_MEM = True

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract dataset (Colab) ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device & dataset checks ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found. Expected: {TRAIN_DIR} and {VAL_DIR}")

# ---------------- Class discovery ----------------
classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

# ---------------- Transforms (with optional RandAugment) ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform_list = [
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
]
if USE_RANDAUG:
    try:
        train_transform_list.append(transforms.RandAugment(num_ops=RANDAUG_N, magnitude=RANDAUG_M))
    except Exception:
        pass

train_transform_list += [
    transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
]
train_transform = transforms.Compose(train_transform_list)

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Datasets & DataLoaders (paths unchanged) ----------------
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

# --- Weighted sampler logic: create sampler only if imbalance detected and enabled ---
y_train = [label for _, label in train_dataset.samples]
class_counts = Counter(y_train)
min_count = min(class_counts.values()) if class_counts else 0
max_count = max(class_counts.values()) if class_counts else 1
use_sampler = False
sampler = None
if USE_WEIGHTED_SAMPLER and len(class_counts) > 1:
    imbalance_ratio = min_count / float(max_count)
    if imbalance_ratio < WEIGHTED_SAMPLER_THRESHOLD:
        use_sampler = True
        print(f"Imbalance detected (min/max = {min_count}/{max_count} -> ratio {imbalance_ratio:.3f}). Using WeightedRandomSampler.")
        weights = [1.0 / class_counts[label] for _, label in train_dataset.samples]
        from torch.utils.data import WeightedRandomSampler
        sampler = WeightedRandomSampler(weights, num_samples=len(weights), replacement=True)
    else:
        print("Imbalance not severe enough; not using sampler.")
else:
    print("Weighted sampler disabled or single class; using default shuffle DataLoader.")

if use_sampler:
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, sampler=sampler,
                                               num_workers=NUM_WORKERS, pin_memory=True)
else:
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                               num_workers=NUM_WORKERS, pin_memory=True)

val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Class weights / Loss choice ----------------
if num_classes > 1 and len(y_train) > 0:
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# Focal loss implementation (optional)
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None, reduction='mean', ignore_index=-100):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
        self.ignore_index = ignore_index

    def forward(self, inputs, targets):
        ce = F.cross_entropy(inputs, targets, reduction='none', ignore_index=self.ignore_index, label_smoothing=0.0)
        p_t = torch.exp(-ce)
        loss = ((1 - p_t) ** self.gamma) * ce
        if self.alpha is not None:
            at = self.alpha[targets].to(inputs.device)
            loss = at * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# Choose base criterion
if USE_FOCAL_LOSS:
    if class_weights is not None:
        alpha = class_weights
    else:
        alpha = None
    criterion = FocalLoss(gamma=FOCAL_GAMMA, alpha=alpha)
else:
    try:
        criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=LABEL_SMOOTHING if LABEL_SMOOTHING>0 else 0.0)
    except TypeError:
        criterion = nn.CrossEntropyLoss(weight=class_weights)

# ---------------- Model (timm Swin) ----------------
print("Creating model:", MODEL_NAME)
model = timm.create_model(MODEL_NAME, pretrained=True, num_classes=num_classes)
model.to(device)
print("Model created:", MODEL_NAME)

# ---------------- Phase 1: freeze backbone, train head ----------------
for name, p in model.named_parameters():
    p.requires_grad = False

# Unfreeze head params heuristically
head_param_count = 0
for name, p in model.named_parameters():
    lname = name.lower()
    if ('head' in lname) or ('classifier' in lname) or lname.endswith('.fc') or lname.endswith('.out_proj'):
        p.requires_grad = True
        head_param_count += 1

if head_param_count == 0:
    for name, p in model.named_parameters():
        if name.endswith('fc') or 'head' in name.lower() or 'classifier' in name.lower():
            p.requires_grad = True
            head_param_count += 1

print("Approx. trainable head parameter groups:", head_param_count)

# ---------------- Parameter groups (fixed: no overlapping params) ----------------
def get_param_groups(model, head_lr=HEAD_LR, backbone_lr_scale=BACKBONE_LR_SCALE, weight_decay=WEIGHT_DECAY):
    """
    Build param groups without duplicates:
      - split into head vs backbone
      - split each into decay vs no_decay (bias and norm)
    Returns a list suitable for AdamW(...)
    """
    head_decay = []
    head_no_decay = []
    backbone_decay = []
    backbone_no_decay = []

    for name, p in model.named_parameters():
        if not p.requires_grad:
            continue

        lname = name.lower()
        # decide if this is a "head" param
        is_head = ('head' in lname) or ('classifier' in lname) or lname.endswith('.fc') or lname.endswith('.out_proj') \
                  or ('head' == lname.split('.')[-1])

        # decide if this param should not use weight decay
        is_no_decay = (p.ndim == 1) or name.endswith(".bias") or ('norm' in lname) or ('bn' in lname) or ('ln' in lname)

        # place param into exactly one list
        if is_head:
            if is_no_decay:
                head_no_decay.append(p)
            else:
                head_decay.append(p)
        else:
            if is_no_decay:
                backbone_no_decay.append(p)
            else:
                backbone_decay.append(p)

    param_groups = []
    if len(head_decay) > 0:
        param_groups.append({"params": head_decay, "lr": head_lr, "weight_decay": weight_decay})
    if len(head_no_decay) > 0:
        param_groups.append({"params": head_no_decay, "lr": head_lr, "weight_decay": 0.0})
    if len(backbone_decay) > 0:
        param_groups.append({"params": backbone_decay, "lr": head_lr * backbone_lr_scale, "weight_decay": weight_decay})
    if len(backbone_no_decay) > 0:
        param_groups.append({"params": backbone_no_decay, "lr": head_lr * backbone_lr_scale, "weight_decay": 0.0})

    return param_groups

# ---------------- Create optimizer for Phase 1 ----------------
param_groups = get_param_groups(model, head_lr=HEAD_LR, backbone_lr_scale=BACKBONE_LR_SCALE, weight_decay=WEIGHT_DECAY)

# Debug check (optional): ensure no duplicates and counts match
param_ids = []
for g in param_groups:
    for p in g["params"]:
        param_ids.append(id(p))
dups = len(param_ids) - len(set(param_ids))
trainable_count = sum(1 for _, p in model.named_parameters() if p.requires_grad)
grouped_count = len(param_ids)
print("Debug param grouping -> duplicates:", dups, "trainable_count:", trainable_count, "grouped_count:", grouped_count)
# If grouped_count < trainable_count, it's okay (some trainable params may be in head/backbone lists not covered by exact heuristics)
# If dups>0, something unexpected happened â€” but the function is designed to avoid duplicates.

optimizer = AdamW(param_groups, lr=HEAD_LR, weight_decay=WEIGHT_DECAY)

# ---------------- Helper: MixUp & CutMix ----------------
def mixup_data(x, y, alpha=MIXUP_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    perm = torch.randperm(x.size(0)).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[perm]
    y_a, y_b = y, y[perm]
    return mixed_x, y_a, y_b, lam

def cutmix_data(x, y, alpha=CUTMIX_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    B, C, H, W = x.shape
    perm = torch.randperm(B).to(x.device)
    cx = np.random.randint(W)
    cy = np.random.randint(H)
    w = int(W * math.sqrt(1 - lam))
    h = int(H * math.sqrt(1 - lam))
    x1 = np.clip(cx - w // 2, 0, W)
    y1 = np.clip(cy - h // 2, 0, H)
    x2 = np.clip(cx + w // 2, 0, W)
    y2 = np.clip(cy + h // 2, 0, H)
    x[:, :, y1:y2, x1:x2] = x[perm, :, y1:y2, x1:x2]
    lam_adj = 1 - ((x2 - x1) * (y2 - y1) / float(W * H))
    y_a, y_b = y, y[perm]
    return x, y_a, y_b, lam_adj

# ---------------- Utilities ----------------
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list)
    y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

# ---------------- Training & validation loops supporting AMP, MixUp, grad accumulation ----------------
scaler = torch.cuda.amp.GradScaler() if (USE_AMP and torch.cuda.is_available()) else None

def train_one_epoch(model, loader, optimizer, scheduler, device, epoch, total_epochs, criterion,
                    grad_accum_steps=1, use_mixup=USE_MIXUP, use_cutmix=USE_CUTMIX, mixup_alpha=MIXUP_ALPHA,
                    cutmix_alpha=CUTMIX_ALPHA, clip_grad_norm=CLIP_GRAD_NORM):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    optimizer.zero_grad()
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    iter_count = 0
    for batch_idx, (imgs, labels) in pbar:
        iter_count += 1
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        # Apply MixUp / CutMix
        if use_cutmix:
            imgs, y_a, y_b, lam = cutmix_data(imgs, labels, alpha=cutmix_alpha)
            use_mixed = True
        elif use_mixup:
            imgs, y_a, y_b, lam = mixup_data(imgs, labels, alpha=mixup_alpha)
            use_mixed = True
        else:
            use_mixed = False

        with torch.cuda.amp.autocast(enabled=(scaler is not None)):
            logits = model(imgs)
            if use_mixed:
                loss = lam * F.cross_entropy(logits, y_a, reduction='mean') + (1.0 - lam) * F.cross_entropy(logits, y_b, reduction='mean')
            else:
                loss = criterion(logits, labels)

        loss = loss / grad_accum_steps
        if scaler is not None:
            scaler.scale(loss).backward()
        else:
            loss.backward()

        if iter_count % grad_accum_steps == 0:
            if scaler is not None:
                scaler.unscale_(optimizer)
                if clip_grad_norm > 0:
                    torch.nn.utils.clip_grad_norm_(model.parameters(), clip_grad_norm)
                scaler.step(optimizer)
                scaler.update()
            else:
                if clip_grad_norm > 0:
                    torch.nn.utils.clip_grad_norm_(model.parameters(), clip_grad_norm)
                optimizer.step()
            optimizer.zero_grad()
            if USE_ONECYCLE:
                try:
                    scheduler.step()
                except Exception:
                    pass

        # bookkeeping
        running_loss += loss.item() * imgs.size(0) * grad_accum_steps
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        b_acc = (preds == labels).float().mean().item()
        running_epoch_loss = running_loss / total if total>0 else 0.0
        running_epoch_acc = correct / total if total>0 else 0.0

        postfix = {
            "b_loss": f"{(loss.item()*grad_accum_steps):.4f}",
            "b_acc": f"{b_acc:.3f}",
            "e_loss": f"{running_epoch_loss:.4f}",
            "e_acc": f"{running_epoch_acc:.3f}"
        }
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)

    pbar.close()
    return running_loss / len(loader.dataset), correct / total if total>0 else 0.0

def validate(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            logits = model(imgs)
            loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset)
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ---------------- Training loop ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS

# Prepare scheduler for Phase 1
if USE_ONECYCLE:
    steps_per_epoch = math.ceil(len(train_loader) / float(GRAD_ACCUM_STEPS))
    try:
        scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=MAX_LR,
                                                        steps_per_epoch=steps_per_epoch,
                                                        epochs=PHASE1_EPOCHS,
                                                        pct_start=0.1, div_factor=10.0, final_div_factor=1e4)
    except Exception as e:
        print("OneCycleLR setup failed (phase1), falling back to ReduceLROnPlateau. Error:", e)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)
else:
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 1 (head-only) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, scheduler, device, epoch, total_epochs, criterion,
                                            grad_accum_steps=GRAD_ACCUM_STEPS, use_mixup=USE_MIXUP, use_cutmix=USE_CUTMIX)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    if not USE_ONECYCLE:
        scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = val_loss
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Phase 2: unfreeze last N swin layers/stages (robust) ----------------
print("Unfreezing last", N_UNFREEZE, "Swin blocks/layers (if available).")
def unfreeze_last_swin_layers(model, n_unfreeze):
    unfrozen = 0
    candidates = []
    if hasattr(model, "layers"):
        candidates.append(("layers", model.layers))
    if hasattr(model, "stages"):
        candidates.append(("stages", model.stages))
    if hasattr(model, "blocks"):
        candidates.append(("blocks", model.blocks))
    if hasattr(model, "features"):
        candidates.append(("features", model.features))
    if hasattr(model, "backbone"):
        backbone = getattr(model, "backbone")
        if hasattr(backbone, "layers"):
            candidates.append(("backbone.layers", backbone.layers))
        if hasattr(backbone, "stages"):
            candidates.append(("backbone.stages", backbone.stages))

    for name, container in candidates:
        try:
            total = len(container)
            start_idx = max(0, total - n_unfreeze)
            print(f"Detected container '{name}' with {total} elements. Unfreezing indices [{start_idx} .. {total-1}].")
            for i in range(start_idx, total):
                block = container[i]
                for p in block.parameters():
                    p.requires_grad = True
                    unfrozen += 1
            if unfrozen > 0:
                return True
        except Exception as e:
            print(f"Candidate '{name}' unfreeze attempt failed: {e}")
            continue

    # fallback: unfreeze entire model
    if unfrozen == 0:
        print("Could not detect logical containers; unfreezing entire model.")
        for name, p in model.named_parameters():
            p.requires_grad = True
        return True
    return unfrozen > 0

try:
    unfreeze_last_swin_layers(model, N_UNFREEZE)
except Exception as e:
    print("Unfreeze fallback error:", e)
    for p in model.parameters():
        p.requires_grad = True

# Recreate optimizer with layer-wise LR before fine-tuning
param_groups = get_param_groups(model, head_lr=HEAD_LR, backbone_lr_scale=BACKBONE_LR_SCALE, weight_decay=WEIGHT_DECAY)

# Debug grouping check again
param_ids = []
for g in param_groups:
    for p in g["params"]:
        param_ids.append(id(p))
dups = len(param_ids) - len(set(param_ids))
trainable_count = sum(1 for _, p in model.named_parameters() if p.requires_grad)
grouped_count = len(param_ids)
print("After unfreeze -> duplicates:", dups, "trainable_count:", trainable_count, "grouped_count:", grouped_count)

optimizer = AdamW(param_groups, lr=HEAD_LR, weight_decay=WEIGHT_DECAY)

# Setup scheduler for Phase 2
if USE_ONECYCLE:
    steps_per_epoch = math.ceil(len(train_loader) / float(GRAD_ACCUM_STEPS))
    try:
        scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=MAX_LR,
                                                        steps_per_epoch=steps_per_epoch,
                                                        epochs=PHASE2_EPOCHS,
                                                        pct_start=0.1, div_factor=10.0, final_div_factor=1e4)
    except Exception as e:
        print("OneCycleLR setup failed (phase2), falling back to ReduceLROnPlateau. Error:", e)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)
else:
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, scheduler, device, epoch, total_epochs, criterion,
                                            grad_accum_steps=GRAD_ACCUM_STEPS, use_mixup=USE_MIXUP, use_cutmix=USE_CUTMIX)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    if not USE_ONECYCLE:
        scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Final evaluation & plotting (unchanged behavior) ----------------
print("Loading best model from:", CHECKPOINT_PATH)
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# ---------------- Save plots ----------------
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(epochs_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, "swin_repl_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot to:", acc_path)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(epochs_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, "swin_repl_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot to:", loss_path)

cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, "swin_repl_confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()
print("Saved confusion matrix to:", cm_path)

plt.figure(figsize=(9,8))
plotted = False
for i in range(num_classes):
    if i in roc_auc_dict:
        plt.plot(fpr_dict[i], tpr_dict[i], label=f"{classes[i]} (AUC={roc_auc_dict[i]:.3f})"); plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title("ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, "swin_repl_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot to:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

# ---------------- Save final model ----------------
torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
