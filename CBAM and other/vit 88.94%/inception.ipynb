```python
# Corrected PyTorch ViT training & evaluation (Colab-ready)
# - Ensures classifier reset happens before moving model to device
# - Uses torch.amp (newer AMP API) for scaler/autocast
# - Robust timm factory + full checkpointing
#
# Requirements (uncomment to install in a fresh Colab runtime):
# !pip install -q torch torchvision timm scikit-learn tqdm matplotlib seaborn

import os
import random
import zipfile
import time
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import transforms, datasets
import timm
from tqdm import tqdm

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# ============== Configuration ==============
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/d/train"
VAL_DIR   = "/content/dataset/d/val"
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_vit_checkpoint.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_vit_state_dict.pt"

REQUESTED_MODEL_NAME = "vit_base_patch4_224"   # primary requested name (factory will fallback)
UPGRADE_TIMM = False  # set True to attempt pip install -U timm (runtime restart required)

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2
SEED = 42
PHASE1_EPOCHS = 8
PHASE2_EPOCHS = 18
N_UNFREEZE = 6
SHOW_GPU_MEM = True
USE_AMP = True
SAVE_FULL_CHECKPOINT = True

# Optional timm upgrade (user must restart runtime afterward)
if UPGRADE_TIMM:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-U", "timm"])
    print("timm upgrade attempted — restart runtime to apply changes.")

# ============== Reproducibility ==============
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ============== Drive mount & dataset ==============
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ============== Device and dataset checks ==============
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found. Expected: {TRAIN_DIR} and {VAL_DIR}")

# ============== Classes, transforms, loaders ==============
classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])
val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

y_train = [label for _, label in train_dataset.samples]
if num_classes > 1 and len(y_train) > 0:
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# ============== timm model factory ==============
def create_timm_vit_model(requested_name, num_classes, pretrained=True, verbose=True):
    available = set(timm.list_models())
    if verbose:
        print(f"[timm] available models count: {len(available)}")
    if requested_name in available:
        if verbose: print(f"[timm] using exact model name: {requested_name}")
        return timm.create_model(requested_name, pretrained=pretrained, num_classes=num_classes)
    # fallbacks
    base_alts = [
        requested_name,
        "vit_base_patch16_224",
        "vit_base_patch32_224",
        "vit_base_patch16_384",
        "vit_base_patch32_384",
        "vit_large_patch16_224",
        "vit_small_patch16_224",
        "vit_base_patch16_224_in21k",
        "vit_base_patch4_224_in21k",
        "vit_base_patch8_224",
    ]
    for alt in base_alts:
        if alt in available:
            if verbose: print(f"[timm] using fallback '{alt}' for requested '{requested_name}'")
            return timm.create_model(alt, pretrained=pretrained, num_classes=num_classes)
    vi_models = [m for m in sorted(available) if "vit" in m]
    if verbose:
        if vi_models:
            print("ViT-like models available (sample):", vi_models[:40])
        else:
            print("No ViT-like models detected in timm.list_models(). Consider upgrading timm.")
    if vi_models:
        fallback = vi_models[0]
        if verbose: print(f"[timm] falling back to '{fallback}'")
        return timm.create_model(fallback, pretrained=pretrained, num_classes=num_classes)
    raise RuntimeError(f"Requested timm model '{requested_name}' not available. Consider `!pip install -U timm` and restart.")

# ============== Create model (reset classifier BEFORE moving to device) ==============
print("Creating model (requested):", REQUESTED_MODEL_NAME)
model = create_timm_vit_model(REQUESTED_MODEL_NAME, num_classes=num_classes, pretrained=True, verbose=True)

# If timm offers reset_classifier, call it BEFORE moving the model to device
try:
    if hasattr(model, "reset_classifier"):
        model.reset_classifier(num_classes=num_classes)
        print("Called model.reset_classifier(...) before moving to device.")
except Exception as e:
    print("reset_classifier attempt failed (continuing):", e)

# Now move model to device (ensures all params/buffers are on same device)
model.to(device)
print("Model moved to device:", device)

# ============== Freeze encoder / unfreeze head heuristic ==============
for name, p in model.named_parameters():
    p.requires_grad = False

head_param_count = 0
head_keywords = ["head", "classifier", "fc", "mlp_head", "head.fc"]
for name, p in model.named_parameters():
    n = name.lower()
    if any(k in n for k in head_keywords) or n.endswith("head"):
        p.requires_grad = True
        head_param_count += 1

if head_param_count == 0:
    params_list = list(model.named_parameters())
    for name, p in params_list[-8:]:
        p.requires_grad = True
        head_param_count += 1

print("Approx. trainable head parameter groups (heuristic):", head_param_count)

# ============== Loss, optimizer, scheduler, AMP scaler ==============
criterion = nn.CrossEntropyLoss(weight=class_weights) if class_weights is not None else nn.CrossEntropyLoss()
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=2e-4, weight_decay=1e-2)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

# New API: torch.amp.GradScaler
scaler = torch.amp.GradScaler(enabled=(USE_AMP and torch.cuda.is_available()))

# ============== Utilities and metrics ==============
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    try:
        return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"
    except Exception:
        return ""

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list)
    y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred) if len(y_true) > 0 else 0.0
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    if n_classes == 0:
        return acc, report, {}, {}, {}, 0.0, []
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = float(np.mean(aucs)) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

# ============== Train & validate helpers (use torch.amp.autocast) ==============
def train_one_epoch(model, loader, optimizer, device, epoch, total_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    for batch_idx, (imgs, labels) in pbar:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        optimizer.zero_grad()
        with torch.amp.autocast(device_type='cuda', enabled=(USE_AMP and torch.cuda.is_available())):
            logits = model(imgs)
            loss = criterion(logits, labels)
        if USE_AMP and torch.cuda.is_available():
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(filter(lambda p: p.requires_grad, model.parameters()), max_norm=5.0)
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            torch.nn.utils.clip_grad_norm_(filter(lambda p: p.requires_grad, model.parameters()), max_norm=5.0)
            optimizer.step()
        running_loss += float(loss.item()) * imgs.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        b_acc = (preds == labels).float().mean().item()
        running_epoch_loss = running_loss / total if total > 0 else 0.0
        running_epoch_acc = correct / total if total > 0 else 0.0
        postfix = {
            "b_loss": f"{loss.item():.4f}",
            "b_acc": f"{b_acc:.3f}",
            "e_loss": f"{running_epoch_loss:.4f}",
            "e_acc": f"{running_epoch_acc:.3f}"
        }
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    return running_loss / len(loader.dataset) if len(loader.dataset) > 0 else 0.0, correct / total if total > 0 else 0.0

def validate(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            with torch.amp.autocast(device_type='cuda', enabled=(USE_AMP and torch.cuda.is_available())):
                logits = model(imgs)
                loss = criterion(logits, labels)
            running_loss += float(loss.item()) * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean() if len(preds) > 0 else 0.0
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels) > 0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels) > 0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset) if len(loader.dataset) > 0 else 0.0
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ============== Training loop with checkpointing ==============
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS

print("Starting Phase 1 (head-only) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val) if len(y_true_val)>0 else 0.0)
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    improved = (val_loss < best_val_loss) or (val_auc_macro > best_auc_macro)
    if improved:
        best_val_loss = val_loss if val_loss < best_val_loss else best_val_loss
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        if SAVE_FULL_CHECKPOINT:
            checkpoint = {
                "epoch": epoch,
                "model_state_dict": model.state_dict(),
                "optimizer_state_dict": optimizer.state_dict(),
                "best_val_loss": best_val_loss,
                "best_auc_macro": best_auc_macro
            }
            if USE_AMP and torch.cuda.is_available():
                checkpoint["scaler_state_dict"] = scaler.state_dict()
            torch.save(checkpoint, CHECKPOINT_PATH)
            print("Saved full checkpoint to:", CHECKPOINT_PATH)
        else:
            torch.save(model.state_dict(), CHECKPOINT_PATH)
            print("Saved model state dict to:", CHECKPOINT_PATH)

# ============== Phase 2: unfreeze last N blocks ==============
print("Unfreezing last", N_UNFREEZE, "transformer blocks (if available).")
try:
    unfreezed = False
    blocks_attr = None
    if hasattr(model, "blocks"):
        blocks_attr = model.blocks
    elif hasattr(model, "transformer") and hasattr(model.transformer, "blocks"):
        blocks_attr = model.transformer.blocks
    elif hasattr(model, "encoder") and hasattr(model.encoder, "layer"):
        blocks_attr = model.encoder.layer

    if blocks_attr is not None:
        total_blocks = len(blocks_attr)
        print("Transformer blocks detected:", total_blocks)
        start_idx = max(0, total_blocks - N_UNFREEZE)
        for i in range(start_idx, total_blocks):
            for p in blocks_attr[i].parameters():
                p.requires_grad = True
        unfreezed = True
    if not unfreezed:
        for p in model.parameters():
            p.requires_grad = True
        print("Fallback: made entire model trainable.")
except Exception as e:
    print("Unfreeze fallback — setting entire model trainable. Error:", e)
    for p in model.parameters():
        p.requires_grad = True

optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-5, weight_decay=1e-2)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val) if len(y_true_val)>0 else 0.0)
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    improved = (val_loss < best_val_loss) or (val_auc_macro > best_auc_macro)
    if improved:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        if SAVE_FULL_CHECKPOINT:
            checkpoint = {
                "epoch": epoch,
                "model_state_dict": model.state_dict(),
                "optimizer_state_dict": optimizer.state_dict(),
                "best_val_loss": best_val_loss,
                "best_auc_macro": best_auc_macro
            }
            if USE_AMP and torch.cuda.is_available():
                checkpoint["scaler_state_dict"] = scaler.state_dict()
            torch.save(checkpoint, CHECKPOINT_PATH)
            print("Saved full checkpoint to:", CHECKPOINT_PATH)
        else:
            torch.save(model.state_dict(), CHECKPOINT_PATH)
            print("Saved model state dict to:", CHECKPOINT_PATH)

# ============== Final evaluation ==============
if os.path.exists(CHECKPOINT_PATH):
    try:
        ckpt = torch.load(CHECKPOINT_PATH, map_location=device)
        if isinstance(ckpt, dict) and "model_state_dict" in ckpt:
            model.load_state_dict(ckpt["model_state_dict"])
            print("Loaded checkpoint model state from:", CHECKPOINT_PATH)
            if "optimizer_state_dict" in ckpt:
                try:
                    optimizer.load_state_dict(ckpt["optimizer_state_dict"])
                    print("Loaded optimizer state from checkpoint.")
                except Exception:
                    print("Optimizer state could not be loaded (incompatible keys).")
            if "scaler_state_dict" in ckpt and USE_AMP and torch.cuda.is_available():
                try:
                    scaler.load_state_dict(ckpt["scaler_state_dict"])
                    print("Loaded AMP scaler state from checkpoint.")
                except Exception:
                    print("AMP scaler state could not be loaded.")
        else:
            model.load_state_dict(ckpt)
            print("Loaded model state dict (legacy) from:", CHECKPOINT_PATH)
    except Exception as e:
        print("Could not load checkpoint cleanly:", e)

model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# ============== Plots & save ==============
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(epochs_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, "vit_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot to:", acc_path)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(epochs_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, "vit_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot to:", loss_path)

if len(y_true_val) > 0:
    cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
    plt.figure(figsize=(7,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
    cm_path = os.path.join(GRAPH_DIR, "vit_confusion_matrix.png")
    plt.savefig(cm_path, dpi=150); plt.show()
    print("Saved confusion matrix to:", cm_path)

plotted = False
plt.figure(figsize=(9,8))
for i in range(num_classes):
    if i in roc_auc_dict:
        plt.plot(fpr_dict[i], tpr_dict[i], label=f"{classes[i]} (AUC={roc_auc_dict[i]:.3f})"); plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title("ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, "vit_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot to:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
