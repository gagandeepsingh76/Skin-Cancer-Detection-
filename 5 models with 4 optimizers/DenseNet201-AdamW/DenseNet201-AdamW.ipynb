```python
# -*- coding: utf-8 -*-
"""desnet_enh_with_adamw.ipynb
DenseNet201 training with selectable optimizer: 'sgd', 'adamax', 'nadam', or 'adamw'
"""

# ---- Mount & dataset extraction (Colab) ----
from google.colab import drive
drive.mount('/content/drive')

import zipfile, os
zip_path = "/content/drive/MyDrive/d.zip"
extract_path = "/content/dataset"

if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("Dataset extracted.")
else:
    print("Dataset already exists at:", extract_path)

# ---- Imports ----
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os, math, collections

from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, accuracy_score
from sklearn.preprocessing import label_binarize

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.densenet import preprocess_input, DenseNet201
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

# Optimizers and regularizers
from tensorflow.keras.optimizers import Adamax, Nadam, SGD
from tensorflow.keras import regularizers

# ---- User settings ----
optimizer_choice = 'adamw'   # choose: 'sgd', 'adamax', 'nadam', 'adamw'

# SGD-specific settings
sgd_phase1_lr = 1e-2      # head-only LR when using SGD
sgd_phase2_lr = 1e-4      # fine-tune LR when using SGD
sgd_momentum = 0.9
sgd_nesterov = True

# Adamax / Nadam defaults
adamax_phase1_lr = 2e-3
adamax_phase2_lr = 1e-4
nadam_phase1_lr = 1e-3
nadam_phase2_lr = 1e-4

# AdamW defaults
adamw_phase1_lr = 3e-4
adamw_phase2_lr = 1e-4
adamw_weight_decay = 1e-5

# Whether to apply L2 regularization to the custom head when using SGD
apply_l2_to_head = True
l2_weight = 1e-4

# Choose phase lrs based on optimizer
opt = optimizer_choice.lower()
if opt == 'adamax':
    phase1_lr = adamax_phase1_lr
    phase2_lr = adamax_phase2_lr
elif opt == 'nadam':
    phase1_lr = nadam_phase1_lr
    phase2_lr = nadam_phase2_lr
elif opt == 'sgd':
    phase1_lr = sgd_phase1_lr
    phase2_lr = sgd_phase2_lr
elif opt == 'adamw':
    phase1_lr = adamw_phase1_lr
    phase2_lr = adamw_phase2_lr
else:
    raise ValueError("optimizer_choice must be 'sgd', 'adamax', 'nadam', or 'adamw'")

print("Optimizer choice:", optimizer_choice)
print(f"Phase1 lr: {phase1_lr}, Phase2 lr: {phase2_lr}")

# ---- Data paths and params ----
train_dir = "/content/dataset/d/train"
val_dir   = "/content/dataset/d/val"
BATCH_SIZE = 32
target_size = (224, 224)
n_classes = 5  # change if different

# ---- Data generators ----
train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

val_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=target_size,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    shuffle=True
)

val_data = val_gen.flow_from_directory(
    val_dir,
    target_size=target_size,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    shuffle=False
)

# ---- Build DenseNet201 base + custom head ----
base_model = DenseNet201(weights='imagenet', include_top=False, input_shape=(target_size[0], target_size[1], 3))

# Freeze all base layers initially; we'll unfreeze last N later
for layer in base_model.layers:
    layer.trainable = False

# custom head: apply L2 to head Dense layers if using SGD and apply_l2_to_head True
head_regularizer = regularizers.l2(l2_weight) if (opt == 'sgd' and apply_l2_to_head) else None

x = GlobalAveragePooling2D()(base_model.output)
x = BatchNormalization()(x)
if head_regularizer:
    x = Dense(256, activation='relu', kernel_regularizer=head_regularizer)(x)
else:
    x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
if head_regularizer:
    outputs = Dense(n_classes, activation='softmax', kernel_regularizer=head_regularizer)(x)
else:
    outputs = Dense(n_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.summary()

# ---- Callbacks ----
checkpoint_phase1 = ModelCheckpoint(
    f"/content/drive/MyDrive/best_densenet201_head_{optimizer_choice}.keras",
    save_best_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1
)

checkpoint_phase2 = ModelCheckpoint(
    f"/content/drive/MyDrive/best_densenet201_finetuned_{optimizer_choice}.keras",
    save_best_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1
)

lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,
    patience=2,
    verbose=1,
    min_lr=1e-6
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

# ---- Training hyperparameters ----
phase1_epochs = 21   # head-only
phase2_epochs = 21   # fine-tune
total_epochs = phase1_epochs + phase2_epochs
N = 50  # last N layers to unfreeze during fine-tuning (tune as needed)

# ---------- PHASE 1: Train head only (backbone frozen) ----------
for layer in base_model.layers:
    layer.trainable = False

# Choose optimizer instance for phase1
if opt == 'adamax':
    optimizer_phase1 = Adamax(learning_rate=phase1_lr)
elif opt == 'nadam':
    optimizer_phase1 = Nadam(learning_rate=phase1_lr)
elif opt == 'sgd':
    optimizer_phase1 = SGD(learning_rate=phase1_lr, momentum=sgd_momentum, nesterov=sgd_nesterov)
elif opt == 'adamw':
    # Try to import AdamW from tf.keras, fallback to tensorflow_addons
    try:
        from tensorflow.keras.optimizers import AdamW
        _adamw_available = True
        print("Using tf.keras.optimizers.AdamW")
    except Exception:
        _adamw_available = False
    if not _adamw_available:
        try:
            import tensorflow_addons as tfa
            AdamW = tfa.optimizers.AdamW
            _adamw_available = True
            print("Using tensorflow_addons.optimizers.AdamW")
        except Exception:
            _adamw_available = False
    if not _adamw_available:
        raise ImportError(
            "AdamW optimizer not found in tf.keras and tensorflow_addons isn't installed.\n"
            "In Colab run: !pip install -q tensorflow-addons\n"
            "Then restart the runtime and re-run this cell."
        )
    # construct AdamW (handle possible API differences)
    try:
        optimizer_phase1 = AdamW(learning_rate=phase1_lr, weight_decay=adamw_weight_decay)
    except TypeError:
        optimizer_phase1 = AdamW(adamw_weight_decay, learning_rate=phase1_lr)
else:
    raise ValueError("Unexpected optimizer choice")

model.compile(
    optimizer=optimizer_phase1,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.build((None, target_size[0], target_size[1], 3))

history_phase1 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=phase1_epochs,
    callbacks=[checkpoint_phase1, lr_scheduler],
    verbose=1
)

# ---------- PHASE 2: Unfreeze last N layers and fine-tune ----------
for layer in base_model.layers[-N:]:
    layer.trainable = True
for layer in base_model.layers[:-N]:
    layer.trainable = False

# Choose optimizer instance for phase2
if opt == 'adamax':
    optimizer_phase2 = Adamax(learning_rate=phase2_lr)
elif opt == 'nadam':
    optimizer_phase2 = Nadam(learning_rate=phase2_lr)
elif opt == 'sgd':
    optimizer_phase2 = SGD(learning_rate=phase2_lr, momentum=sgd_momentum, nesterov=sgd_nesterov)
elif opt == 'adamw':
    # AdamW instance for phase2 (reuse same import logic)
    try:
        from tensorflow.keras.optimizers import AdamW
        _adamw_available = True
    except Exception:
        _adamw_available = False
    if not _adamw_available:
        try:
            import tensorflow_addons as tfa
            AdamW = tfa.optimizers.AdamW
            _adamw_available = True
        except Exception:
            _adamw_available = False
    if not _adamw_available:
        raise ImportError(
            "AdamW optimizer not found in tf.keras and tensorflow_addons isn't installed.\n"
            "In Colab run: !pip install -q tensorflow-addons\n"
            "Then restart the runtime and re-run this cell."
        )
    try:
        optimizer_phase2 = AdamW(learning_rate=phase2_lr, weight_decay=adamw_weight_decay)
    except TypeError:
        optimizer_phase2 = AdamW(adamw_weight_decay, learning_rate=phase2_lr)
else:
    raise ValueError("Unexpected optimizer choice")

model.compile(
    optimizer=optimizer_phase2,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_phase2 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=total_epochs,
    initial_epoch=phase1_epochs,
    callbacks=[checkpoint_phase2, lr_scheduler, early_stopping],
    verbose=1
)

# ---- Plotting results ----
graph_dir = "/content/drive/MyDrive/diabetic_eye_results"
os.makedirs(graph_dir, exist_ok=True)

histories = []
try:
    histories.append(history_phase1.history)
except NameError:
    pass
try:
    histories.append(history_phase2.history)
except NameError:
    pass

if not histories:
    try:
        histories.append(history.history)
    except NameError:
        raise RuntimeError("No training history found.")

def concat_metric(metric_name):
    concatenated = []
    for h in histories:
        if metric_name in h:
            concatenated.extend(h[metric_name])
    return concatenated

train_acc = concat_metric('accuracy')
val_acc   = concat_metric('val_accuracy')
train_loss = concat_metric('loss')
val_loss   = concat_metric('val_loss')

num_epochs = len(train_acc)
epochs = np.arange(1, num_epochs + 1)

# Accuracy
plt.figure(figsize=(8,5))
plt.plot(epochs, train_acc, marker='o', label='Train Accuracy')
if val_acc:
    plt.plot(epochs, val_acc, marker='o', label='Validation Accuracy')
plt.title("Accuracy over Epochs (DenseNet201)")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.grid(True, linestyle='--', alpha=0.4); plt.legend(loc='best')
if val_acc:
    best_idx = int(np.argmax(val_acc))
    best_val = val_acc[best_idx]
    plt.annotate(f"Best val_acc = {best_val:.4f} (epoch {best_idx+1})",
                 xy=(best_idx+1, best_val),
                 xytext=(best_idx+1, best_val - 0.07 if best_val > 0.2 else best_val + 0.02),
                 arrowprops=dict(arrowstyle="->", lw=0.8))
acc_path = os.path.join(graph_dir, f"densenet201_accuracy_plot_{optimizer_choice}.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()

# Loss
plt.figure(figsize=(8,5))
plt.plot(epochs, train_loss, marker='o', label='Train Loss')
if val_loss:
    plt.plot(epochs, val_loss, marker='o', label='Validation Loss')
plt.title("Loss over Epochs (DenseNet201)")
plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.grid(True, linestyle='--', alpha=0.4); plt.legend(loc='best')
if val_loss:
    best_loss_idx = int(np.argmin(val_loss))
    best_loss = val_loss[best_loss_idx]
    plt.annotate(f"Lowest val_loss = {best_loss:.4f} (epoch {best_loss_idx+1})",
                 xy=(best_loss_idx+1, best_loss),
                 xytext=(best_loss_idx+1, best_loss + 0.05 if best_loss < 1.0 else best_loss - 0.05),
                 arrowprops=dict(arrowstyle="->", lw=0.8))
loss_path = os.path.join(graph_dir, f"densenet201_loss_plot_{optimizer_choice}.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()

print(f"Saved accuracy plot to: {acc_path}")
print(f"Saved loss plot to:     {loss_path}")

# ---- Load best model for evaluation (prefer finetuned) ----
preferred_paths = [
    f"/content/drive/MyDrive/best_densenet201_finetuned_{optimizer_choice}.keras",
    f"/content/drive/MyDrive/best_densenet201_head_{optimizer_choice}.keras"
]

model = None
for p in preferred_paths:
    if os.path.exists(p):
        try:
            model = load_model(p, compile=False)
            model_path = p
            print("Loaded model from:", p)
            break
        except Exception as e:
            print("Warning: failed to load model at", p, ":", e)

if model is None:
    raise FileNotFoundError(f"No checkpoint found in: {preferred_paths}")

# ---- Evaluation ----
graph_dir = "/content/drive/MyDrive/diabetic_eye_results"
os.makedirs(graph_dir, exist_ok=True)

val_samples = val_data.samples
batch_size = val_data.batch_size
steps = int(math.ceil(val_samples / float(batch_size)))

val_data.reset()
pred_probs = model.predict(val_data, steps=steps, verbose=1)
y_pred = np.argmax(pred_probs, axis=1)
if len(y_pred) > val_samples:
    y_pred = y_pred[:val_samples]

y_true = val_data.classes
if len(y_true) > val_samples:
    y_true = y_true[:val_samples]

class_indices = val_data.class_indices
index_to_label = {v:k for k,v in class_indices.items()}
valid_labels = sorted(index_to_label.keys())
target_names = [index_to_label[i] for i in valid_labels]

print("Unique y_true:", np.unique(y_true))
print("Unique y_pred:", np.unique(y_pred))
print("Target names:", target_names)

# Predicted counts
print("\nPredicted sample count per class:")
pred_counter = collections.Counter(y_pred)
for label in valid_labels:
    print(f"{index_to_label[label]}: {pred_counter.get(label, 0)}")

accuracy = accuracy_score(y_true, y_pred)
print(f"\nOverall Validation Accuracy: {accuracy * 100:.2f}%")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=valid_labels)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix (DenseNet201)")
plt.tight_layout()
conf_mat_path = os.path.join(graph_dir, f"densenet201_confusion_matrix_{optimizer_choice}.png")
plt.savefig(conf_mat_path, dpi=150); plt.show()
print("Saved confusion matrix to:", conf_mat_path)

# Classification report
report = classification_report(y_true, y_pred, labels=valid_labels, target_names=target_names, zero_division=0)
print("\nClassification Report:\n"); print(report)
report_path = os.path.join(graph_dir, f"densenet201_classification_report_{optimizer_choice}.txt")
with open(report_path, "w") as f:
    f.write(f"Model path: {model_path}\n")
    f.write(f"Validation samples: {val_samples}\n\n")
    f.write(report)
print("Saved classification report to:", report_path)

# ROC curves
labels_plot = [f"Class {i}" for i in range(n_classes)]
roc_dir = os.path.join(graph_dir, "roc")
os.makedirs(roc_dir, exist_ok=True)

y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
fpr, tpr, roc_auc = {}, {}, {}
missing_classes = []

plt.figure(figsize=(10,8))
for i in range(n_classes):
    if np.sum(y_true_bin[:, i]) == 0:
        print(f"Skipping ROC for class {i} — no positive samples.")
        missing_classes.append(i)
        continue
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'{labels_plot[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0,1],[0,1],'k--',label='Chance')
plt.title('ROC Curve (one-vs-rest)')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.legend(loc='lower right'); plt.grid(True); plt.tight_layout()
roc_path = os.path.join(roc_dir, f"roc_curve_{optimizer_choice}.png")
plt.savefig(roc_path); plt.show()
if missing_classes:
    print("Skipped ROC for indices with no positives:", missing_classes)
print("ROC curve saved to:", roc_path)

# ---- Save final model ----
final_model_path = f"/content/drive/MyDrive/final_densenet201_model_{optimizer_choice}.keras"
model.save(final_model_path)
print("Saved final model to:", final_model_path)
```
