```python
# Full Colab cell: CNN training + evaluation with optimizer choice (sgd | adamax | adamw)
# Paste into a single Colab cell and run.

# ---------------------------
# Configuration / Settings
# ---------------------------
OPTIMIZER = "adamw"   # choose: "sgd", "adamax", or "adamw"
BATCH_SIZE = 32
IMG_SIZE = (128, 128)
NUM_CLASSES = 5        # set to your dataset's number of classes
EPOCHS = 51
SEED = 42

# Paths (Drive)
DRIVE_MOUNT_POINT = "/content/drive"
ZIP_PATH = "/content/drive/MyDrive/d.zip"   # your zip on Drive
DATA_ROOT = "/content/dataset"
TRAIN_DIR = os.path.join(DATA_ROOT, "d", "train")
VAL_DIR = os.path.join(DATA_ROOT, "d", "val")

# Output artifact names (will include optimizer name)
CHECKPOINT_PATH = f"/content/drive/MyDrive/best_cnn_model_{OPTIMIZER}.keras"
FINAL_MODEL_PATH = f"/content/drive/MyDrive/final_cnn_model_{OPTIMIZER}.keras"
GRAPHS_DIR = f"/content/drive/MyDrive/diabetic_eye_results_{OPTIMIZER}"

# Optimizer hyperparams
LR_SGD = 0.01
MOMENTUM = 0.9
LR_ADAMAX = 0.002
LR_ADAMW  = 0.001
WEIGHT_DECAY = 1e-5
CLIPNORM = None   # set e.g. 1.0 if you want gradient clipping

# ---------------------------
# Imports & mount Drive
# ---------------------------
import os, zipfile, collections
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize

# Mount Drive (Colab)
from google.colab import drive
drive.mount(DRIVE_MOUNT_POINT, force_remount=False)

# ---------------------------
# Optional: seed & mixed precision
# ---------------------------
np.random.seed(SEED)
tf.random.set_seed(SEED)

# If you want mixed precision on GPU uncomment (careful with some TF builds)
# try:
#     if tf.config.list_physical_devices('GPU'):
#         tf.keras.mixed_precision.set_global_policy("mixed_float16")
#         print("Mixed precision enabled")
# except Exception as e:
#     print("Mixed precision not enabled:", e)

# ---------------------------
# Extract dataset if needed
# ---------------------------
if not os.path.exists(DATA_ROOT):
    os.makedirs(DATA_ROOT, exist_ok=True)

if os.path.exists(ZIP_PATH) and not (os.path.exists(TRAIN_DIR) and os.path.exists(VAL_DIR)):
    print("Extracting dataset zip from Drive to", DATA_ROOT)
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(DATA_ROOT)
    print("Extraction complete.")
else:
    print("Dataset already present or zip not found. DATA_ROOT =", DATA_ROOT)

# ---------------------------
# Data generators / loaders
# ---------------------------
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    shuffle=True,
    seed=SEED
)

val_data = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    shuffle=False
)

# Update NUM_CLASSES from generator if mismatch
NUM_CLASSES = len(train_data.class_indices)
print("Detected classes:", train_data.class_indices)

# ---------------------------
# Model definition (simple CNN)
# ---------------------------
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

def build_cnn(input_shape=(*IMG_SIZE, 3), n_classes=NUM_CLASSES):
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2,2)),
        Conv2D(64, (3,3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2,2)),
        Conv2D(128, (3,3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2,2)),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        BatchNormalization(),
        Dense(n_classes, activation='softmax')
    ])
    return model

model = build_cnn()
model.build((None, *IMG_SIZE, 3))
model.summary()

# ---------------------------
# Optimizer selection (sgd | adamax | adamw)
# ---------------------------
from tensorflow.keras.optimizers import SGD

optimizer = None
opt_lower = OPTIMIZER.lower().strip()

if opt_lower == "sgd":
    optimizer = SGD(learning_rate=LR_SGD, momentum=MOMENTUM, nesterov=True, clipnorm=CLIPNORM)
    print("Selected optimizer: SGD")
elif opt_lower == "adamax":
    from tensorflow.keras.optimizers import Adamax
    optimizer = Adamax(learning_rate=LR_ADAMAX, beta_1=0.9, beta_2=0.999, epsilon=1e-7, clipnorm=CLIPNORM)
    print("Selected optimizer: Adamax")
elif opt_lower == "adamw":
    # Try standard tf.keras.optimizers.AdamW, else try experimental path, else fallback to Adamax
    try:
        from tensorflow.keras.optimizers import AdamW as TFAdamW
        optimizer = TFAdamW(learning_rate=LR_ADAMW, weight_decay=WEIGHT_DECAY, beta_1=0.9, beta_2=0.999, epsilon=1e-7, clipnorm=CLIPNORM)
        print("Selected optimizer: tf.keras.optimizers.AdamW")
    except Exception as e1:
        try:
            # newer TF may expose experimental AdamW separately
            from tensorflow.keras.optimizers.experimental import AdamW as TFExpAdamW
            optimizer = TFExpAdamW(learning_rate=LR_ADAMW, weight_decay=WEIGHT_DECAY, beta_1=0.9, beta_2=0.999, epsilon=1e-7, clipnorm=CLIPNORM)
            print("Selected optimizer: tf.keras.optimizers.experimental.AdamW (fallback)")
        except Exception as e2:
            from tensorflow.keras.optimizers import Adamax
            optimizer = Adamax(learning_rate=LR_ADAMAX, beta_1=0.9, beta_2=0.999, epsilon=1e-7, clipnorm=CLIPNORM)
            print("WARNING: AdamW not available in this TF build; falling back to Adamax.")
            print("Errors:", e1, e2)
else:
    raise ValueError("Unsupported optimizer: choose 'sgd', 'adamax', or 'adamw'")

# ---------------------------
# Compile model
# ---------------------------
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ---------------------------
# Callbacks: checkpoint, lr reduce, earlystop
# ---------------------------
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

os.makedirs(GRAPHS_DIR, exist_ok=True)

checkpoint_cb = ModelCheckpoint(
    CHECKPOINT_PATH,
    save_best_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1
)

reduce_lr_cb = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,
    patience=2,
    verbose=1,
    min_lr=1e-6
)

earlystop_cb = EarlyStopping(
    monitor='val_loss',
    patience=6,
    restore_best_weights=True,
    verbose=1
)

# ---------------------------
# Train
# ---------------------------
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS,
    callbacks=[checkpoint_cb, reduce_lr_cb, earlystop_cb],
    verbose=1
)

# ---------------------------
# Save final model (attempt .keras then fallback .h5)
# ---------------------------
try:
    model.save(FINAL_MODEL_PATH)
    print("Saved final model to:", FINAL_MODEL_PATH)
except Exception as e:
    print("Saving .keras failed, attempting .h5 fallback:", e)
    try:
        fallback = FINAL_MODEL_PATH.replace(".keras", ".h5")
        model.save(fallback)
        print("Saved fallback model to:", fallback)
    except Exception as e2:
        print("Final save also failed:", e2)

# ---------------------------
# Plot accuracy & loss
# ---------------------------
plt.figure(); plt.plot(history.history.get('accuracy', []), label='Train Acc'); plt.plot(history.history.get('val_accuracy', []), label='Val Acc')
plt.title("Accuracy"); plt.legend(); plt.grid(True)
plt.savefig(os.path.join(GRAPHS_DIR, "accuracy_plot.png")); plt.show()

plt.figure(); plt.plot(history.history.get('loss', []), label='Train Loss'); plt.plot(history.history.get('val_loss', []), label='Val Loss')
plt.title("Loss"); plt.legend(); plt.grid(True)
plt.savefig(os.path.join(GRAPHS_DIR, "loss_plot.png")); plt.show()

# ---------------------------
# Load best checkpoint / evaluate on validation
# ---------------------------
from tensorflow.keras.models import load_model
best_model_path = CHECKPOINT_PATH if os.path.exists(CHECKPOINT_PATH) else FINAL_MODEL_PATH
print("Attempting to load best model from:", best_model_path)
try:
    model = load_model(best_model_path)
    print("Loaded model:", best_model_path)
except Exception as e:
    print("Could not load checkpoint; continuing with in-memory model. Error:", e)

# Ensure generator ordering & predict
val_data.reset()
pred_probs = model.predict(val_data, verbose=1)
y_pred = np.argmax(pred_probs, axis=1)
y_true = val_data.classes

# Class mapping
class_indices = val_data.class_indices
index_to_label = {v: k for k, v in class_indices.items()}
valid_labels = sorted(index_to_label.keys())
target_names = [index_to_label[i] for i in valid_labels]

# Accuracy & confusion matrix
acc = accuracy_score(y_true, y_pred)
print(f"\nValidation accuracy: {acc*100:.2f}%")
cm = confusion_matrix(y_true, y_pred, labels=valid_labels)

plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix")
plt.savefig(os.path.join(GRAPHS_DIR, "confusion_matrix.png")); plt.show()

print("\nClassification report:\n")
print(classification_report(y_true, y_pred, labels=valid_labels, target_names=target_names, zero_division=0))

# ---------------------------
# ROC per-class (one-vs-rest)
# ---------------------------
n_classes = NUM_CLASSES
labels_for_roc = list(range(n_classes))
y_true_bin = label_binarize(y_true, classes=labels_for_roc)
if pred_probs.shape[1] != n_classes:
    # If model output shape mismatches expected classes, try to handle it
    print("Warning: predicted probability shape", pred_probs.shape, "does not match NUM_CLASSES", n_classes)

fpr, tpr, roc_auc = {}, {}, {}
missing = []
for i in range(n_classes):
    if np.sum(y_true_bin[:, i]) == 0:
        print(f"Skipping ROC for class {i} ({index_to_label.get(i,'?')}) — no positives in y_true")
        missing.append(i)
        continue
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(9,8))
for i in range(n_classes):
    if i in roc_auc:
        plt.plot(fpr[i], tpr[i], label=f"{index_to_label.get(i, i)} (AUC={roc_auc[i]:.3f})")
plt.plot([0,1],[0,1],'k--', label='Chance')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC Curves (one-vs-rest)")
plt.legend(loc='lower right'); plt.grid(True)
plt.savefig(os.path.join(GRAPHS_DIR, "roc_curve.png")); plt.show()

if missing:
    print("Skipped ROC for classes:", [index_to_label.get(i, i) for i in missing])
print("ROC figure saved to:", os.path.join(GRAPHS_DIR, "roc_curve.png"))

# ---------------------------
# Save prediction CSV (optional)
# ---------------------------
import pandas as pd
rows = []
filenames = val_data.filenames  # generator filenames align with predictions
for f, true, pred, probs in zip(filenames, y_true, y_pred, pred_probs):
    row = {"filename": f, "true_label": index_to_label[int(true)], "pred_label": index_to_label[int(pred)]}
    # optionally include per-class probabilities
    for ci in range(pred_probs.shape[1]):
        row[f"prob_class_{ci}"] = float(probs[ci])
    rows.append(row)
df = pd.DataFrame(rows)
csv_out = os.path.join(GRAPHS_DIR, f"val_predictions_{OPTIMIZER}.csv")
df.to_csv(csv_out, index=False)
print("Saved per-sample predictions to:", csv_out)

print("\nAll done. Artifacts stored under:", GRAPHS_DIR)
```
