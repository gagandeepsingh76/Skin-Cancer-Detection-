```python
# -*- coding: utf-8 -*-
"""inceptionv3_enh_with_adamw.ipynb
"""

# Mount Drive & extract dataset (Colab)
from google.colab import drive
drive.mount('/content/drive')

import zipfile, os, sys
zip_path = "/content/drive/MyDrive/d.zip"
extract_path = "/content/dataset"

if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("Dataset extracted.")
else:
    print("Dataset already exists at:", extract_path)

# ---- Imports ----
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
import tensorflow as tf

from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, accuracy_score
from sklearn.preprocessing import label_binarize

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Optional: set random seed for reproducibility (not fully deterministic across TF/CuDNN versions)
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---- Paths and params ----
train_dir = "/content/dataset/d/train"
val_dir = "/content/dataset/d/val"
batch_size = 32
target_size = (299, 299)
n_classes = 5  # change if different
epochs = 51

# ---- Data augmentation / generators ----
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

val_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=True,
    seed=SEED
)

val_data = val_gen.flow_from_directory(
    val_dir,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)

# ---- Build model (InceptionV3 base + custom head) ----
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(target_size[0], target_size[1], 3))

# Freeze early layers (you can adjust number)
freeze_until = 200
for i, layer in enumerate(base_model.layers):
    layer.trainable = False if i < freeze_until else True

# Classification head
x = GlobalAveragePooling2D()(base_model.output)
x = BatchNormalization()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
outputs = Dense(n_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)

# Print model summary (optional)
model.summary()

# ---- AdamW optimizer (tf.keras or tensorflow_addons fallback) ----
try:
    # TF 2.11+ typically has this
    from tensorflow.keras.optimizers import AdamW
    adamw_available = True
    print("Using tf.keras.optimizers.AdamW")
except Exception:
    adamw_available = False

if not adamw_available:
    try:
        import tensorflow_addons as tfa
        # tensorflow_addons has tfa.optimizers.AdamW
        AdamW = tfa.optimizers.AdamW
        adamw_available = True
        print("Using tensorflow_addons.optimizers.AdamW")
    except Exception:
        adamw_available = False

if not adamw_available:
    raise ImportError(
        "AdamW optimizer not found in tf.keras and tensorflow_addons isn't installed.\n"
        "In Colab run: !pip install -q tensorflow-addons\n"
        "Then restart the runtime and re-run this cell."
    )

# Hyperparameters for AdamW -- tune these
adamw_lr = 1e-4
weight_decay = 1e-5
# Some AdamW implementations expect weight_decay as first/keyword arg. Both tf.keras and tfa accept weight_decay keyword.

# Construct optimizer instance depending on implementation specifics
try:
    # tf.keras.optimizers.AdamW and tfa.optimizers.AdamW both accept weight_decay and learning_rate keywords
    optimizer = AdamW(learning_rate=adamw_lr, weight_decay=weight_decay)
except TypeError:
    # Some older tfa versions accepted weight_decay as first arg
    optimizer = AdamW(weight_decay, learning_rate=adamw_lr)

# Optional: gradient clipping (uncomment if you want)
# optimizer = tf.keras.optimizers.Adam(learning_rate=adamw_lr, clipnorm=1.0)

# ---- Compile model ----
model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Build model input (optional, helpful for some TF versions)
model.build((None, target_size[0], target_size[1], 3))

# ---- Callbacks ----
checkpoint_path = "/content/drive/MyDrive/best_inceptionv3_model.keras"
os.makedirs(os.path.dirname(checkpoint_path), exist_ok=True)

checkpoint = ModelCheckpoint(
    checkpoint_path,
    save_best_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1
)

lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,
    patience=2,
    verbose=1,
    min_lr=1e-6
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

callbacks = [checkpoint, lr_scheduler, early_stopping]

# ---- Train ----
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=epochs,
    callbacks=callbacks
)

# ---- Create results directory ----
graph_dir = "/content/drive/MyDrive/diabetic_eye_results"
os.makedirs(graph_dir, exist_ok=True)

# ---- Plot Accuracy ----
plt.figure()
plt.plot(history.history.get('accuracy', []), label='Train Accuracy')
plt.plot(history.history.get('val_accuracy', []), label='Validation Accuracy')
plt.legend()
plt.title("Accuracy over Epochs")
plt.savefig(f"{graph_dir}/accuracy_plot.png")
plt.show()

# ---- Plot Loss ----
plt.figure()
plt.plot(history.history.get('loss', []), label='Train Loss')
plt.plot(history.history.get('val_loss', []), label='Validation Loss')
plt.legend()
plt.title("Loss over Epochs")
plt.savefig(f"{graph_dir}/loss_plot.png")
plt.show()

# ---- Load best checkpoint for evaluation (if saved) ----
if os.path.exists(checkpoint_path):
    try:
        model = load_model(checkpoint_path)
        print("Loaded best model from checkpoint:", checkpoint_path)
    except Exception as e:
        print("Warning: could not load checkpoint (maybe custom objects). Using current model instance. Error:", e)

# ---- Predictions on validation set ----
# Ensure generator is reset
val_data.reset()
pred_probs = model.predict(val_data, verbose=1)
y_pred = np.argmax(pred_probs, axis=1)
y_true = val_data.classes

# ---- Class mapping ----
class_indices = val_data.class_indices
index_to_label = {v: k for k, v in class_indices.items()}
valid_labels = sorted(index_to_label.keys())
target_names = [index_to_label[i] for i in valid_labels]

# Debug prints
print("Unique y_true:", np.unique(y_true))
print("Unique y_pred:", np.unique(y_pred))
print("Target names:", target_names)

# Predicted counts per class
print("\nPredicted sample count per class:")
pred_counter = collections.Counter(y_pred)
for label in valid_labels:
    print(f"{index_to_label[label]}: {pred_counter.get(label, 0)}")

# ---- Accuracy ----
accuracy = accuracy_score(y_true, y_pred)
print(f"\nOverall Validation Accuracy: {accuracy * 100:.2f}%")

# ---- Confusion Matrix ----
cm = confusion_matrix(y_true, y_pred, labels=valid_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.savefig(f"{graph_dir}/confusion_matrix.png")
plt.show()

# ---- Classification Report ----
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, labels=valid_labels, target_names=target_names, zero_division=0))

# ---- ROC curves (one-vs-rest) ----
os.makedirs(f"{graph_dir}/roc", exist_ok=True)

# Binarize true labels for ROC computation
y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))

fpr, tpr, roc_auc = {}, {}, {}
missing_classes = []

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    if np.sum(y_true_bin[:, i]) == 0:
        print(f"Skipping ROC for class index {i} ({index_to_label.get(i, 'Unknown')}) — no positive samples.")
        missing_classes.append(i)
        continue
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'{index_to_label.get(i, i)} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Chance')
plt.title('ROC Curve (one-vs-rest)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
roc_path = os.path.join(graph_dir, "roc_curve.png")
plt.savefig(roc_path)
plt.show()
if missing_classes:
    print(f"Skipped ROC for indices with no positive samples: {missing_classes}")
print("ROC curve saved to:", roc_path)

# ---- Final save (full model) ----
final_model_path = "/content/drive/MyDrive/final_inceptionv3_model.keras"
model.save(final_model_path)
print("Saved final model to:", final_model_path)
```
