```python
# -*- coding: utf-8 -*-
"""inceptionv3_with_optimizer_choice.ipynb
Complete pipeline. Choose optimizer by setting optimizer_choice = 'sgd' or 'adamw'.
"""

# ---- Mount Drive & dataset extraction (Colab) ----
from google.colab import drive
drive.mount('/content/drive')

import zipfile, os, sys
zip_path = "/content/drive/MyDrive/d.zip"
extract_path = "/content/dataset"

if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("Dataset extracted.")
else:
    print("Dataset already exists at:", extract_path)

# ---- Imports ----
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
import tensorflow as tf

from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, accuracy_score
from sklearn.preprocessing import label_binarize

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# ---- Reproducibility ----
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---- USER SETTINGS ----
optimizer_choice = 'sgd'   # 'sgd' or 'adamw'
# ---------- SGD settings ----------
sgd_lr = 0.01
sgd_momentum = 0.9
sgd_nesterov = True        # set False if you don't want Nesterov
# Optionally apply L2 regularization to custom head to mimic "weight decay" with SGD
apply_l2_to_head = True
l2_weight = 1e-4           # strength for kernel_regularizer in the head

# ---------- AdamW settings ----------
adamw_lr = 1e-4
adamw_weight_decay = 1e-5

# ---- Paths and params ----
train_dir = "/content/dataset/d/train"
val_dir = "/content/dataset/d/val"
batch_size = 32
target_size = (299, 299)
n_classes = 5  # change if different
epochs = 51

# ---- Data augmentation / generators ----
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

val_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    shear_range=0.15,
    fill_mode='nearest'
)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=True,
    seed=SEED
)

val_data = val_gen.flow_from_directory(
    val_dir,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)

# ---- Build model (InceptionV3 base + custom head) ----
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(target_size[0], target_size[1], 3))

# Freeze early layers (tune as needed)
freeze_until = 200
for i, layer in enumerate(base_model.layers):
    layer.trainable = False if i < freeze_until else True

# Create custom head. If using SGD + L2, add kernel_regularizer only to head Dense layers.
from tensorflow.keras import regularizers
head_regularizer = regularizers.l2(l2_weight) if (optimizer_choice == 'sgd' and apply_l2_to_head) else None

x = GlobalAveragePooling2D()(base_model.output)
x = BatchNormalization()(x)
if head_regularizer:
    x = Dense(256, activation='relu', kernel_regularizer=head_regularizer)(x)
else:
    x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
if head_regularizer:
    outputs = Dense(n_classes, activation='softmax', kernel_regularizer=head_regularizer)(x)
else:
    outputs = Dense(n_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.summary()

# ---- Select optimizer based on choice ----
if optimizer_choice.lower() == 'adamw':
    # Try tf.keras AdamW; fallback to tensorflow_addons
    try:
        from tensorflow.keras.optimizers import AdamW
        print("Using tf.keras.optimizers.AdamW")
    except Exception:
        try:
            import tensorflow_addons as tfa
            AdamW = tfa.optimizers.AdamW
            print("Using tensorflow_addons.optimizers.AdamW")
        except Exception:
            raise ImportError(
                "AdamW not found. In Colab, run: !pip install -q tensorflow-addons\n"
                "Then restart runtime and re-run."
            )
    # Construct AdamW optimizer (most implementations accept learning_rate & weight_decay)
    try:
        optimizer = AdamW(learning_rate=adamw_lr, weight_decay=adamw_weight_decay)
    except TypeError:
        optimizer = AdamW(adamw_weight_decay, learning_rate=adamw_lr)
    optimizer_name = f"AdamW_lr{adamw_lr}_wd{adamw_weight_decay}"

elif optimizer_choice.lower() == 'sgd':
    from tensorflow.keras.optimizers import SGD
    optimizer = SGD(learning_rate=sgd_lr, momentum=sgd_momentum, nesterov=sgd_nesterov)
    optimizer_name = f"SGD_lr{sgd_lr}_mom{sgd_momentum}_nesterov{sgd_nesterov}"

else:
    raise ValueError("optimizer_choice must be 'sgd' or 'adamw'")

print("Using optimizer:", optimizer_name)

# ---- Compile model ----
model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.build((None, target_size[0], target_size[1], 3))

# ---- Callbacks ----
graph_dir = "/content/drive/MyDrive/diabetic_eye_results"
os.makedirs(graph_dir, exist_ok=True)

checkpoint_path = f"/content/drive/MyDrive/best_inceptionv3_{optimizer_choice}.keras"
os.makedirs(os.path.dirname(checkpoint_path), exist_ok=True)

checkpoint = ModelCheckpoint(
    checkpoint_path,
    save_best_only=True,
    monitor='val_loss',
    mode='min',
    verbose=1
)

lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,
    patience=2,
    verbose=1,
    min_lr=1e-6
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

callbacks = [checkpoint, lr_scheduler, early_stopping]

# ---- Train ----
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=epochs,
    callbacks=callbacks
)

# ---- Plots (accuracy & loss) ----
os.makedirs(graph_dir, exist_ok=True)

plt.figure()
plt.plot(history.history.get('accuracy', []), label='Train Accuracy')
plt.plot(history.history.get('val_accuracy', []), label='Validation Accuracy')
plt.legend(); plt.title("Accuracy over Epochs")
plt.savefig(f"{graph_dir}/accuracy_plot_{optimizer_choice}.png")
plt.show()

plt.figure()
plt.plot(history.history.get('loss', []), label='Train Loss')
plt.plot(history.history.get('val_loss', []), label='Validation Loss')
plt.legend(); plt.title("Loss over Epochs")
plt.savefig(f"{graph_dir}/loss_plot_{optimizer_choice}.png")
plt.show()

# ---- Load best checkpoint if exists ----
if os.path.exists(checkpoint_path):
    try:
        model = load_model(checkpoint_path)
        print("Loaded best model from checkpoint:", checkpoint_path)
    except Exception as e:
        print("Warning: could not load checkpoint (maybe custom objects). Using current model instance. Error:", e)

# ---- Evaluation on validation set ----
val_data.reset()
pred_probs = model.predict(val_data, verbose=1)
y_pred = np.argmax(pred_probs, axis=1)
y_true = val_data.classes

class_indices = val_data.class_indices
index_to_label = {v: k for k, v in class_indices.items()}
valid_labels = sorted(index_to_label.keys())
target_names = [index_to_label[i] for i in valid_labels]

print("Unique y_true:", np.unique(y_true))
print("Unique y_pred:", np.unique(y_pred))
print("Target names:", target_names)

print("\nPredicted sample count per class:")
pred_counter = collections.Counter(y_pred)
for label in valid_labels:
    print(f"{index_to_label[label]}: {pred_counter.get(label, 0)}")

accuracy = accuracy_score(y_true, y_pred)
print(f"\nOverall Validation Accuracy: {accuracy * 100:.2f}%")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=valid_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=target_names, yticklabels=target_names)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix")
plt.tight_layout()
plt.savefig(f"{graph_dir}/confusion_matrix_{optimizer_choice}.png")
plt.show()

# Classification report
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, labels=valid_labels, target_names=target_names, zero_division=0))

# ROC curves
os.makedirs(f"{graph_dir}/roc", exist_ok=True)
y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
fpr, tpr, roc_auc = {}, {}, {}
missing_classes = []

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    if np.sum(y_true_bin[:, i]) == 0:
        print(f"Skipping ROC for class index {i} ({index_to_label.get(i, 'Unknown')}) — no positive samples.")
        missing_classes.append(i)
        continue
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'{index_to_label.get(i, i)} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Chance')
plt.title('ROC Curve (one-vs-rest)')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.legend(loc='lower right'); plt.grid(True)
plt.tight_layout()
roc_path = os.path.join(graph_dir, f"roc_curve_{optimizer_choice}.png")
plt.savefig(roc_path)
plt.show()

if missing_classes:
    print("Skipped ROC for indices with no positive samples:", missing_classes)

print("ROC curve saved to:", roc_path)

# ---- Final save ----
final_model_path = f"/content/drive/MyDrive/final_inceptionv3_model_{optimizer_choice}.keras"
model.save(final_model_path)
print("Saved final model to:", final_model_path)
```
