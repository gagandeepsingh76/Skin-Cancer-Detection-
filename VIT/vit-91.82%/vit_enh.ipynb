```python
# Enhanced ViT training pipeline with multiple transfer-learning techniques
# Keep dataset paths unchanged.
# Requirements: torch, torchvision, timm, scikit-learn, tqdm, matplotlib, seaborn
# Optionally install in Colab:
# !pip install -q torch torchvision timm scikit-learn tqdm matplotlib seaborn

import os
import re
import random
import zipfile
import time
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import OneCycleLR, CosineAnnealingLR, ReduceLROnPlateau
from torch.optim.swa_utils import AveragedModel, SWALR
from torchvision import transforms, datasets
import timm
from tqdm import tqdm

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight
from torch.utils.data.sampler import WeightedRandomSampler

# ---------------- Configuration (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"            # dataset zip on Drive (if used)
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/d/train"
VAL_DIR   = "/content/dataset/d/val"
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_vit_plus.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_vit_plus.pt"

# Model selection: try preferred, fallback if OOM or not available
PREFERRED_MODEL = "vit_base_patch16_224"   # try "vit_large_patch16_224" if GPU >=16GB
FALLBACK_MODEL = "efficientnet_b3"         # or "resnet50"
IMG_SIZE = 224

# Training hyperparameters (sensible defaults)
BATCH_SIZE = 24
NUM_WORKERS = 4
SEED = 42
PHASE1_EPOCHS = 8    # head-only
PHASE2_EPOCHS = 24   # fine-tune
N_UNFREEZE = 8       # number of transformer blocks to unfreeze (ViT)
SHOW_GPU_MEM = True

# Advanced technique toggles / values
USE_MIXUP = True
MIXUP_ALPHA = 0.2
USE_CUTMIX = True
CUTMIX_ALPHA = 1.0
CUTMIX_PROB = 0.4
LABEL_SMOOTHING = 0.05
USE_FOCAL_LOSS = False          # set True if extreme class imbalance / many hard examples
FOCAL_GAMMA = 2.0
LAYER_DECAY = 0.75              # for layer-wise lr: earlier layers lr = base_lr * layer_decay^(depth - layer_idx)
HEAD_LR = 3e-4
FINETUNE_LR = 3e-5
WEIGHT_DECAY = 1e-2
MAX_GRAD_NORM = 1.0
ACCUMULATION_STEPS = 1
USE_ONECYCLE = True
USE_COSINE_FINETUNE = False
USE_SWA = True
SWA_START_EPOCH = PHASE1_EPOCHS + max(1, PHASE2_EPOCHS - 6)
EARLY_STOPPING_PATIENCE = 8
TTA_TRANSFORMS = [
    lambda x: x,
    lambda x: torch.flip(x, dims=[-1]),
]

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract dataset (Colab) ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device & dataset checks ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found. Expected: {TRAIN_DIR} and {VAL_DIR}")

# ---------------- Class discovery ----------------
classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

# ---------------- Transforms (strong augmentations) ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform = transforms.Compose([
    transforms.Resize(int(IMG_SIZE * 1.12)),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.6, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(25),
    transforms.ColorJitter(brightness=0.18, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    transforms.RandomErasing(p=0.2, scale=(0.01, 0.12)),
])

val_transform = transforms.Compose([
    transforms.Resize(int(IMG_SIZE * 1.12)),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Datasets & DataLoaders ----------------
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

# Weighted sampler for class-balancing (optional)
y_train = [label for _, label in train_dataset.samples]
if num_classes > 1 and len(y_train) > 0:
    class_counts = np.bincount(y_train, minlength=num_classes)
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    sample_weights = np.array([class_weights_np[y] for y in y_train], dtype=np.float32)
    sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)
    print("Class counts:", class_counts.tolist())
    print("Class weights:", class_weights_np)
else:
    class_weights = None
    sampler = None

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE,
                                           sampler=sampler if sampler is not None else None,
                                           shuffle=(sampler is None),
                                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Model creation with fallback ----------------
def create_model(preferred, fallback, pretrained=True, num_classes=2):
    try:
        m = timm.create_model(preferred, pretrained=pretrained, num_classes=num_classes)
        print("Created model:", preferred)
        return m, preferred
    except Exception as e:
        print(f"Preferred model {preferred} creation failed ({e}). Falling back to {fallback}.")
        m = timm.create_model(fallback, pretrained=pretrained, num_classes=num_classes)
        print("Created fallback model:", fallback)
        return m, fallback

model, used_model_name = create_model(PREFERRED_MODEL, FALLBACK_MODEL, pretrained=True, num_classes=num_classes)
model.to(device)

# ---------------- Replace classifier head with dropout+linear head ----------------
def replace_head_with_dropout(m, num_classes, dropout=0.4):
    # Identify head name and replace with Sequential(Dropout, Linear). Works for many timm models.
    found = False
    for attr in ["head", "classifier", "fc", "head.fc", "head.fc.weight"]:
        if hasattr(m, "head") and isinstance(getattr(m, "head"), nn.Module):
            # common pattern for ViT timm: m.head -> nn.Linear
            in_features = None
            # try to deduce in_features
            head = m.head
            if isinstance(head, nn.Linear):
                in_features = head.in_features
                m.head = nn.Sequential(nn.Dropout(dropout), nn.Linear(in_features, num_classes))
                found = True
                break
            elif isinstance(head, nn.Sequential):
                # replace final linear
                for i,layer in enumerate(head):
                    if isinstance(layer, nn.Linear):
                        in_features = layer.in_features
                        head[i] = nn.Sequential(nn.Dropout(dropout), nn.Linear(in_features, num_classes))
                        m.head = head
                        found = True
                        break
                if found: break
    # Some timm models use .classifier
    if not found and hasattr(m, "classifier"):
        cl = m.classifier
        if isinstance(cl, nn.Linear):
            in_features = cl.in_features
            m.classifier = nn.Sequential(nn.Dropout(dropout), nn.Linear(in_features, num_classes))
            found = True
    # If still not found, try to append a new head attribute
    if not found:
        # fallback: add attribute 'head_soft' and leave original intact
        embed_dim = None
        # try common attributes
        for name, p in m.named_parameters():
            if name.endswith("norm.weight"):
                # approximate embed dim from preceding layer param shape
                embed_dim = p.shape[0]
                break
        if embed_dim is None:
            # guess 1024
            embed_dim = 1024
        m.head = nn.Sequential(nn.Dropout(dropout), nn.Linear(embed_dim, num_classes))
    return m

model = replace_head_with_dropout(model, num_classes=num_classes, dropout=0.4)
model.to(device)

# ---------------- Freeze encoder (phase1) ----------------
for n, p in model.named_parameters():
    p.requires_grad = False

# Unfreeze classifier head params
head_param_count = 0
for name, p in model.named_parameters():
    if 'head' in name or 'classifier' in name or name.endswith('.fc'):
        p.requires_grad = True
        head_param_count += 1
# fallback unfreeze last layers if none found
if head_param_count == 0:
    for name, p in list(model.named_parameters())[-40:]:
        p.requires_grad = True

print("Approx. trainable head parameter groups:", head_param_count)

# ---------------- Loss functions ----------------
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, weight=None, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.weight = weight
        self.reduction = reduction

    def forward(self, input, target):
        logpt = -F.cross_entropy(input, target, weight=self.weight, reduction='none', label_smoothing=0.0)
        pt = logpt.exp()
        loss = -((1 - pt) ** self.gamma) * logpt
        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        else:
            return loss

def get_criterion():
    if USE_FOCAL_LOSS:
        return FocalLoss(gamma=FOCAL_GAMMA, weight=class_weights)
    else:
        try:
            return nn.CrossEntropyLoss(weight=class_weights, label_smoothing=LABEL_SMOOTHING)
        except TypeError:
            # older torch fallback
            return nn.CrossEntropyLoss(weight=class_weights)

criterion = get_criterion()

# ---------------- MixUp / CutMix ----------------
def mixup_data(x, y, alpha=MIXUP_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    batch_size = x.size()[0]
    if batch_size == 1:
        return x, y, None, 1.0
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def cutmix_data(x, y, alpha=CUTMIX_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    batch_size, c, h, w = x.size()
    index = torch.randperm(batch_size).to(x.device)
    cut_rat = np.sqrt(1. - lam)
    cut_w = int(w * cut_rat)
    cut_h = int(h * cut_rat)
    cx = np.random.randint(w); cy = np.random.randint(h)
    x1 = np.clip(cx - cut_w // 2, 0, w); y1 = np.clip(cy - cut_h // 2, 0, h)
    x2 = np.clip(cx + cut_w // 2, 0, w); y2 = np.clip(cy + cut_h // 2, 0, h)
    x[:, :, y1:y2, x1:x2] = x[index, :, y1:y2, x1:x2]
    lam_adj = 1. - ((x2 - x1) * (y2 - y1) / float(w * h))
    y_a, y_b = y, y[index]
    return x, y_a, y_b, lam_adj

def mixup_criterion(pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# ---------------- Parameter grouping for layer-wise lr decay (ViT) ----------------
def get_layer_id_for_vit(param_name):
    # ViT parameters usually have names like 'blocks.0.attn.qkv.weight' or 'patch_embed.proj.weight'
    m = re.search(r'blocks\.(\d+)\.', param_name)
    if m:
        return int(m.group(1)) + 1  # 1-based layer id for blocks
    if param_name.startswith('patch_embed') or 'pos_embed' in param_name:
        return 0
    # classifier/head etc.
    return 1000  # head group high id

def get_parameter_groups(model, base_lr, weight_decay, layer_decay=LAYER_DECAY):
    # Build param groups: earlier layers get smaller lr by factor layer_decay^(depth - layer_id)
    param_groups = {}
    # find max layer id
    layer_ids = []
    for n, p in model.named_parameters():
        lid = get_layer_id_for_vit(n)
        layer_ids.append(lid)
    max_id = max(layer_ids + [0])
    for n, p in model.named_parameters():
        if not p.requires_grad:
            continue
        lid = get_layer_id_for_vit(n)
        # compute scale
        depth = max_id if max_id > 0 else 1
        scale = layer_decay ** (depth - lid)
        group_name = f"layer_{lid}"
        if group_name not in param_groups:
            param_groups[group_name] = {"params": [], "lr": base_lr * scale, "weight_decay": weight_decay}
        param_groups[group_name]["params"].append(p)
    return list(param_groups.values())

# ---------------- Utilities ----------------
scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())

def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list); y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

# ---------------- Training & Validation routines (with AMP, MixUp/CutMix, TTA) ----------------
def tta_predict_batch(model, imgs):
    model.eval()
    probs_sum = None
    with torch.no_grad():
        for tfn in TTA_TRANSFORMS:
            aug = tfn(imgs)
            logits = model(aug)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            probs_sum = probs if probs_sum is None else probs_sum + probs
    return probs_sum / len(TTA_TRANSFORMS)

def train_one_epoch(model, loader, optimizer, epoch, total_epochs, scheduler=None):
    model.train()
    running_loss = 0.0; correct = 0; total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    for batch_idx, (imgs, labels) in pbar:
        imgs = imgs.to(device); labels = labels.to(device)
        # augmentation strategy
        use_cutmix = USE_CUTMIX and (np.random.rand() < CUTMIX_PROB)
        if use_cutmix:
            imgs_m, y_a, y_b, lam = cutmix_data(imgs.clone(), labels, alpha=CUTMIX_ALPHA)
        elif USE_MIXUP:
            imgs_m, y_a, y_b, lam = mixup_data(imgs, labels, alpha=MIXUP_ALPHA)
        else:
            imgs_m, y_a, y_b, lam = imgs, labels, None, 1.0

        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            logits = model(imgs_m)
            if y_b is None:
                loss = criterion(logits, labels)
            else:
                loss = mixup_criterion(logits, y_a, y_b, lam)
            loss = loss / ACCUMULATION_STEPS

        scaler.scale(loss).backward()
        if MAX_GRAD_NORM is not None:
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), MAX_GRAD_NORM)
        scaler.step(optimizer); scaler.update(); optimizer.zero_grad()
        if scheduler is not None and isinstance(scheduler, OneCycleLR):
            scheduler.step()

        running_loss += loss.item() * imgs.size(0) * ACCUMULATION_STEPS
        preds = logits.detach().argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        running_epoch_loss = running_loss / total if total>0 else running_loss
        running_epoch_acc = correct / total if total>0 else 0.0
        postfix = {"b_loss": f"{loss.item()*ACCUMULATION_STEPS:.4f}", "e_loss": f"{running_epoch_loss:.4f}", "e_acc": f"{running_epoch_acc:.4f}"}
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    avg_loss = running_loss / len(loader.dataset) if len(loader.dataset)>0 else 0.0
    return avg_loss, correct / total if total>0 else 0.0

def validate(model, loader, epoch=None):
    model.eval()
    running_loss = 0.0; all_preds=[]; all_labels=[]; all_probs=[]
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device); labels = labels.to(device)
            probs = tta_predict_batch(model, imgs)  # numpy array
            preds = np.argmax(probs, axis=1)
            # For loss compute, forward pass
            with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
                logits = model(imgs)
                loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            all_probs.append(probs); all_preds.extend(list(preds)); all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {"b_loss": f"{loss.item():.4f}", "b_acc": f"{batch_acc:.3f}", "e_loss": f"{running_epoch_loss:.4f}", "e_acc": f"{running_epoch_acc:.3f}"}
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset) if len(loader.dataset)>0 else 0.0
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ---------------- Training loop with discriminative LR, OneCycle and SWA ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_score = -1e9
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS
stopping_counter = 0

# ---------- Phase 1: train head only (OneCycle recommended) ----------
# build parameter groups for only-trainable params (head)
head_params = [p for n,p in model.named_parameters() if p.requires_grad]
optimizer = AdamW(head_params, lr=HEAD_LR, weight_decay=WEIGHT_DECAY)
if USE_ONECYCLE:
    steps_per_epoch = max(1, len(train_loader))
    scheduler = OneCycleLR(optimizer, max_lr=HEAD_LR, total_steps=PHASE1_EPOCHS * steps_per_epoch)
else:
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 1 (head-only) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, epoch, total_epochs, scheduler=scheduler if USE_ONECYCLE else None)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    if not USE_ONECYCLE:
        scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    combined = -val_loss + val_auc_macro
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if combined > best_score:
        best_score = combined
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH); stopping_counter = 0
    else:
        stopping_counter += 1
        print(f"No improvement, early-stop counter: {stopping_counter}/{EARLY_STOPPING_PATIENCE}")
    if stopping_counter >= EARLY_STOPPING_PATIENCE:
        print("Early stopping triggered in Phase 1."); break

# ---------- Phase 2: unfreeze and discriminative LR ----------
print("Unfreezing last", N_UNFREEZE, "blocks (if available) or making whole model trainable.")
try:
    if hasattr(model, "blocks"):
        total_blocks = len(model.blocks)
        print("Transformer blocks detected:", total_blocks)
        for i in range(max(0, total_blocks - N_UNFREEZE), total_blocks):
            for p in model.blocks[i].parameters():
                p.requires_grad = True
    else:
        for p in model.parameters():
            p.requires_grad = True
except Exception as e:
    print("Unfreeze fallback:", e)
    for p in model.parameters():
        p.requires_grad = True

# build parameter groups using layer-wise decay if ViT-like
trainable_params = [ (n,p) for n,p in model.named_parameters() if p.requires_grad ]
if "vit" in used_model_name.lower():
    param_groups = get_parameter_groups(model, base_lr=FINETUNE_LR, weight_decay=WEIGHT_DECAY, layer_decay=LAYER_DECAY)
else:
    # simple single group for CNN backbones
    param_groups = [{"params":[p for n,p in trainable_params], "lr": FINETUNE_LR, "weight_decay": WEIGHT_DECAY}]

optimizer = AdamW(param_groups, lr=FINETUNE_LR, weight_decay=WEIGHT_DECAY)
if USE_COSINE_FINETUNE:
    scheduler = CosineAnnealingLR(optimizer, T_max=PHASE2_EPOCHS)
else:
    # keep ReduceLROnPlateau as robust fallback or OneCycle across whole finetune
    if USE_ONECYCLE:
        steps_per_epoch = max(1, len(train_loader))
        scheduler = OneCycleLR(optimizer, max_lr=FINETUNE_LR, total_steps=PHASE2_EPOCHS * steps_per_epoch)
    else:
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

# SWA setup
swa_model = AveragedModel(model) if USE_SWA else None
swa_scheduler = SWALR(optimizer, swa_lr=FINETUNE_LR*0.5) if USE_SWA else None

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
stopping_counter = 0
for epoch in range(PHASE1_EPOCHS + 1, PHASE1_EPOCHS + PHASE2_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, epoch, total_epochs, scheduler=(scheduler if USE_ONECYCLE else None))
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    if (not USE_ONECYCLE) and isinstance(scheduler, ReduceLROnPlateau):
        scheduler.step(val_loss)
    elif USE_COSINE_FINETUNE and isinstance(scheduler, CosineAnnealingLR):
        scheduler.step()
    if USE_SWA and epoch >= SWA_START_EPOCH:
        swa_model.update_parameters(model); swa_scheduler.step()

    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    combined = -val_loss + val_auc_macro
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")

    if combined > best_score:
        best_score = combined
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH); stopping_counter = 0
    else:
        stopping_counter += 1
        print(f"No improvement, early-stop counter: {stopping_counter}/{EARLY_STOPPING_PATIENCE}")
    if stopping_counter >= EARLY_STOPPING_PATIENCE:
        print("Early stopping triggered in Phase 2."); break

# Finalize SWA and evaluate
if USE_SWA and swa_model is not None:
    print("Finalizing SWA - updating BatchNorm stats.")
    torch.optim.swa_utils.update_bn(train_loader, swa_model)
    final_model = swa_model.module if hasattr(swa_model, 'module') else swa_model
else:
    final_model = model

print("Loading best saved state:", CHECKPOINT_PATH)
final_model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
final_model.to(device); final_model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(final_model, val_loader, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# Save plots & final model
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(epochs_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, "enhanced_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(epochs_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, "enhanced_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()

cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, "enhanced_confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()

torch.save(final_model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
print("Saved plots to:", GRAPH_DIR)
```
