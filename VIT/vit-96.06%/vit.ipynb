```python
# Updated: InceptionV3 -> PyTorch ViT training & evaluation (Colab-ready)
# Adds: RandAugment, MixUp/CutMix, AMP + gradient accumulation, discriminative LRs (warmup + cosine),
# label smoothing, dropout in head, and optional SAM (toggleable).
# Requirements: torch, torchvision, timm, scikit-learn, tqdm, matplotlib, seaborn
# Uncomment to install in a fresh Colab runtime:
# !pip install -q torch torchvision timm scikit-learn tqdm matplotlib seaborn

import os
import random
import zipfile
import time
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import LambdaLR
from torchvision import transforms, datasets
import timm
from tqdm import tqdm

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# ---------------- Configuration (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/edited.zip"            # dataset zip on Drive (if used)
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/edited/train"
VAL_DIR   = "/content/dataset/edited/val"
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_vit_inception_replacement.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_vit_inception_replacement.pt"

MODEL_NAME = "vit_base_patch16_224"   # timm ViT model
IMG_SIZE = 224                        # ViT standard
BATCH_SIZE = 32
NUM_WORKERS = 2                       # safe default for Colab
SEED = 42
PHASE1_EPOCHS = 8
PHASE2_EPOCHS = 18
N_UNFREEZE = 6                        # number of transformer blocks to unfreeze in phase 2
SHOW_GPU_MEM = True                   # show GPU memory in postfix if CUDA available

# ----- New training/regularization options -----
ACCUM_STEPS = 2               # gradient accumulation to simulate larger batch
MIXUP_ALPHA = 0.2             # set >0 to enable MixUp
CUTMIX_ALPHA = 0.0            # set >0 to enable CutMix (MixUp and CutMix are mutually exclusive at batch time)
LABEL_SMOOTHING = 0.1
WEIGHT_DECAY = 1e-2
HEAD_DROPOUT = 0.5

# Learning rates: discriminative setup (head vs unfrozen blocks)
HEAD_LR_PHASE1 = 2e-4
BACKBONE_LR_PHASE1 = 0.0     # backbone frozen in phase1
HEAD_LR_PHASE2 = 5e-5
UNFROZEN_BLOCK_LR = 1e-5
BACKBONE_LR_PHASE2 = 1e-6    # minimal LR for remaining backbone if kept trainable

WARMUP_EPOCHS = 3            # linear warmup for first N epochs then cosine decay
USE_SAM = False              # toggle Sharpness-Aware Minimization (costly ~2x steps). If True, AMP is disabled for reliability.

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract dataset (Colab) ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device & dataset checks ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found. Expected: {TRAIN_DIR} and {VAL_DIR}")

# ---------------- Class discovery ----------------
classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

# ---------------- Transforms (RandAugment + baseline transforms) ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

# RandAugment available in torchvision >= 0.10; if not available fallback to ColorJitter/RandomErasing
try:
    randaug = transforms.RandAugment(num_ops=2, magnitude=9)
except Exception:
    print("RandAugment not available in torchvision; continuing without it.")
    randaug = transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02)

train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    randaug,
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Datasets & DataLoaders ----------------
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Class weights ----------------
y_train = [label for _, label in train_dataset.samples]
if num_classes > 1 and len(y_train) > 0:
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# ---------------- Model (timm ViT) ----------------
# We'll add a small dropout MLP head for extra regularization (wrap timm's classifier if exists)
model = timm.create_model(MODEL_NAME, pretrained=True, num_classes=num_classes)
# Replace the head with a small dropout MLP if classifier exists
def replace_head_with_dropout(model, in_features=None, num_classes=num_classes, dropout=HEAD_DROPOUT):
    # Try common attribute names
    replaced = False
    head_in = None
    for attr in ["head", "fc", "classifier"]:
        if hasattr(model, attr):
            m = getattr(model, attr)
            # try to introspect final linear
            if isinstance(m, nn.Linear):
                head_in = m.in_features
            elif hasattr(m, "in_features"):
                head_in = getattr(m, "in_features")
            elif hasattr(m, "__dict__") and "in_features" in m.__dict__:
                head_in = m.__dict__.get("in_features")
            # Create new head
            if head_in is not None:
                new_head = nn.Sequential(
                    nn.Dropout(dropout),
                    nn.Linear(head_in, max(512, head_in//2)),
                    nn.ReLU(inplace=True),
                    nn.Dropout(dropout),
                    nn.Linear(max(512, head_in//2), num_classes)
                )
                setattr(model, attr, new_head)
                replaced = True
                break
    return replaced

_ = replace_head_with_dropout(model, num_classes=num_classes, dropout=HEAD_DROPOUT)

model.to(device)
print("Model created:", MODEL_NAME)

# ---------------- Phase 1: freeze ViT encoder, train head ----------------
for name, p in model.named_parameters():
    p.requires_grad = False

# Make classifier/head parameters trainable (timm usually uses 'head' naming)
head_param_count = 0
for name, p in model.named_parameters():
    if ("head" in name or "classifier" in name or name.endswith("fc") or "mlp" in name) and p.requires_grad == False:
        p.requires_grad = True
        head_param_count += 1

# fallback: if zero trainable head params, unfreeze layers that look like classifier
if head_param_count == 0:
    for name, p in model.named_parameters():
        if name.endswith("fc") or "head" in name or "classifier" in name:
            p.requires_grad = True
            head_param_count += 1

print("Approx. trainable head parameter groups:", head_param_count)

# ---------------- Loss, optimizer & scheduler helpers ----------------
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, smoothing=0.0, weight=None):
        super().__init__()
        self.smoothing = smoothing
        self.weight = weight
    def forward(self, pred, target):
        if self.smoothing <= 0:
            return F.cross_entropy(pred, target, weight=self.weight)
        with torch.no_grad():
            n = pred.size(1)
            true_dist = torch.zeros_like(pred)
            true_dist.fill_(self.smoothing / (n - 1))
            true_dist.scatter_(1, target.data.unsqueeze(1), 1.0 - self.smoothing)
        log_prob = F.log_softmax(pred, dim=1)
        return torch.mean(torch.sum(- true_dist * log_prob, dim=1))

criterion = LabelSmoothingCrossEntropy(smoothing=max(0.0, LABEL_SMOOTHING), weight=class_weights)

# Scheduler: linear warmup for WARMUP_EPOCHS then cosine decay across remaining epochs
def make_epoch_scheduler(optimizer, total_epochs, warmup_epochs=WARMUP_EPOCHS):
    def lr_lambda(epoch):
        if epoch < warmup_epochs:
            return float(epoch + 1) / float(max(1, warmup_epochs))
        # cosine decay from 1.0 down to 0.0 across remaining epochs
        progress = float(epoch - warmup_epochs) / float(max(1, total_epochs - warmup_epochs))
        return 0.5 * (1.0 + np.cos(np.pi * progress))
    return LambdaLR(optimizer, lr_lambda)

# Build optimizer for phase1 (only head trainable)
def get_param_groups_phase1(model):
    head_params = []
    other_params = []
    for name, p in model.named_parameters():
        if p.requires_grad:
            head_params.append(p)
        else:
            other_params.append(p)
    groups = [
        {"params": head_params, "lr": HEAD_LR_PHASE1, "weight_decay": WEIGHT_DECAY}
    ]
    return groups

# Build optimizer for phase2 (discriminative: head, unfrozen blocks, rest)
def get_param_groups_phase2(model, unfrozen_blocks_names=None):
    # heuristics: head / classifier -> head group
    head_params = []
    unfrozen_params = []
    backbone_params = []
    for name, p in model.named_parameters():
        if not p.requires_grad:
            continue
        lname = name.lower()
        if "head" in lname or "classifier" in lname or "mlp" in lname or "fc" in lname:
            head_params.append(p)
        elif unfrozen_blocks_names and any(nb in name for nb in unfrozen_blocks_names):
            unfrozen_params.append(p)
        else:
            # remaining trainable params (maybe backbone small lr)
            backbone_params.append(p)
    groups = []
    if head_params:
        groups.append({"params": head_params, "lr": HEAD_LR_PHASE2, "weight_decay": WEIGHT_DECAY})
    if unfrozen_params:
        groups.append({"params": unfrozen_params, "lr": UNFROZEN_BLOCK_LR, "weight_decay": WEIGHT_DECAY})
    if backbone_params:
        groups.append({"params": backbone_params, "lr": BACKBONE_LR_PHASE2, "weight_decay": WEIGHT_DECAY})
    return groups

# Simple SAM implementation (supports non-AMP mode). If USE_SAM True, AMP will be disabled to avoid complexity.
class SAM:
    def __init__(self, params, base_optimizer, rho=0.05, adaptive=False, **kwargs):
        self.base_optimizer = base_optimizer(params, **kwargs)
        self.param_groups = self.base_optimizer.param_groups
        self.rho = rho
        self.adaptive = adaptive

    @staticmethod
    def _grad_norm(params, adaptive):
        device = params[0].grad.device
        norm = torch.norm(
            torch.stack([
                ((p.abs() if adaptive else 1.0) * p.grad).norm(p=2).to(device)
                for p in params if p.grad is not None
            ]), p=2)
        return norm

    def first_step(self, zero_grad=False):
        params = [p for group in self.param_groups for p in group['params'] if p.grad is not None]
        grad_norm = SAM._grad_norm(params, self.adaptive)
        scale = self.rho / (grad_norm + 1e-12)
        for group in self.param_groups:
            for p in group['params']:
                if p.grad is None:
                    continue
                e_w = (p.abs() if self.adaptive else 1.0) * p.grad * scale
                p.add_(e_w)               # ascent step
                p.state['e_w'] = e_w
        if zero_grad:
            self.zero_grad()

    def second_step(self, zero_grad=False):
        # revert the ascent step
        for group in self.param_groups:
            for p in group['params']:
                if 'e_w' in p.state:
                    p.sub_(p.state['e_w'])
        self.base_optimizer.step()
        if zero_grad:
            self.zero_grad()

    def zero_grad(self):
        self.base_optimizer.zero_grad()

# ---------------- MixUp / CutMix helpers ----------------
def rand_bbox(size, lam):
    W = size[2]
    H = size[3]
    cut_rat = np.sqrt(1. - lam)
    cut_w = np.int(W * cut_rat)
    cut_h = np.int(H * cut_rat)
    # uniform
    cx = np.random.randint(W)
    cy = np.random.randint(H)
    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)
    return bbx1, bby1, bbx2, bby2

def mixup_data(x, y, alpha=MIXUP_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    index = torch.randperm(x.size(0)).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def cutmix_data(x, y, alpha=CUTMIX_ALPHA):
    if alpha <= 0:
        return x, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    bbx1, bby1, bbx2, bby2 = rand_bbox(x.size(), lam)
    x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]
    # adjust lambda to exactly match pixel ratio
    lam_adjusted = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (x.size(-1) * x.size(-2)))
    y_a, y_b = y, y[index]
    return x, y_a, y_b, lam_adjusted

# ---------------- Utilities ----------------
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list)
    y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

# ---------------- Single-line progress training & validation (with AMP + accumulation + MixUp/CutMix + optional SAM) ----------------
def train_one_epoch(model, loader, optimizer, scaler, device, epoch, total_epochs, amp_enabled=True, accum_steps=1, use_sam=False, sam_obj=None):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    optimizer.zero_grad()
    for step, (imgs, labels) in pbar:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        # Choose augmentation at batch-time (prefer CutMix if configured; else MixUp)
        if CUTMIX_ALPHA > 0:
            imgs_mixed, y_a, y_b, lam = cutmix_data(imgs.clone(), labels, alpha=CUTMIX_ALPHA)
        elif MIXUP_ALPHA > 0:
            imgs_mixed, y_a, y_b, lam = mixup_data(imgs.clone(), labels, alpha=MIXUP_ALPHA)
        else:
            imgs_mixed, y_a, y_b, lam = imgs, labels, labels, 1.0

        if use_sam:
            # SAM + AMP can be complicated; when SAM enabled we disable AMP for reliability
            amp_local = False
        else:
            amp_local = amp_enabled

        if amp_local:
            with torch.cuda.amp.autocast(enabled=True):
                logits = model(imgs_mixed)
                if lam != 1.0:
                    loss = lam * criterion(logits, y_a) + (1. - lam) * criterion(logits, y_b)
                else:
                    loss = criterion(logits, y_a)
                loss = loss / accum_steps
            scaler.scale(loss).backward()
        else:
            logits = model(imgs_mixed)
            if lam != 1.0:
                loss = lam * criterion(logits, y_a) + (1. - lam) * criterion(logits, y_b)
            else:
                loss = criterion(logits, y_a)
            loss = loss / accum_steps
            loss.backward()

        # SAM two-step handling
        if use_sam:
            # First ascent step
            sam_obj.first_step(zero_grad=True)
            # Second forward-backward
            logits2 = model(imgs_mixed)
            if lam != 1.0:
                loss2 = lam * criterion(logits2, y_a) + (1. - lam) * criterion(logits2, y_b)
            else:
                loss2 = criterion(logits2, y_a)
            loss2 = loss2 / accum_steps
            loss2.backward()
            sam_obj.second_step(zero_grad=True)
        else:
            if (step % accum_steps) == 0:
                if amp_local:
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    optimizer.step()
                optimizer.zero_grad()

        # tracking metrics (use logits without grad)
        with torch.no_grad():
            preds = logits.argmax(dim=1)
            running_loss += (loss.item() * accum_steps) * imgs.size(0)  # scaled back to per-sample
            correct += (preds == labels).sum().item()
            total += imgs.size(0)
            b_acc = (preds == labels).float().mean().item()
            running_epoch_loss = running_loss / total if total > 0 else 0.0
            running_epoch_acc = correct / total if total > 0 else 0.0
            postfix = {
                "b_loss": f"{loss.item()*accum_steps:.4f}",
                "b_acc": f"{b_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)

    pbar.close()
    return running_loss / len(loader.dataset), correct / total if total>0 else 0.0

def validate(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            logits = model(imgs)
            loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset)
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ---------------- Training loop ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS

# Prepare phase1 optimizer & scheduler
param_groups_p1 = get_param_groups_phase1(model)
optimizer = AdamW(param_groups_p1, lr=HEAD_LR_PHASE1, weight_decay=WEIGHT_DECAY)
scheduler = make_epoch_scheduler(optimizer, total_epochs=PHASE1_EPOCHS, warmup_epochs=WARMUP_EPOCHS)

# SAM: if enabled, build wrapper (note: SAM disables AMP in this script)
sam_obj = None
if USE_SAM:
    print("SAM enabled: AMP will be disabled for reliability. Expect ~2x cost.")
    # wrap the underlying optimizer (requires passing parameters again)
    base_optimizer_cls = AdamW
    sam_obj = SAM(params=[p for group in param_groups_p1 for p in group["params"]],
                  base_optimizer=base_optimizer_cls, rho=0.05, adaptive=False, lr=HEAD_LR_PHASE1, weight_decay=WEIGHT_DECAY)
    # replace optimizer variable in training to use sam_obj.base_optimizer where needed
    optimizer_for_step = sam_obj.base_optimizer
else:
    optimizer_for_step = optimizer

# AMP scaler (disabled if SAM)
scaler = torch.cuda.amp.GradScaler(enabled=(torch.cuda.is_available() and not USE_SAM))

print("Starting Phase 1 (head-only) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer_for_step, scaler, device,
                                            epoch, total_epochs, amp_enabled=(not USE_SAM), accum_steps=ACCUM_STEPS,
                                            use_sam=USE_SAM, sam_obj=sam_obj)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)

    # scheduler step (epoch-based lambda scheduler)
    scheduler.step()

    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = val_loss
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Phase 2: unfreeze last N transformer blocks ----------------
print("Unfreezing last", N_UNFREEZE, "transformer blocks (if available).")
unfrozen_block_names = []
try:
    # timm ViT exposes blocks as model.blocks or model.transformer.blocks depending on implementation
    if hasattr(model, "blocks"):
        total_blocks = len(model.blocks)
        print("Transformer blocks detected (model.blocks):", total_blocks)
        start = max(0, total_blocks - N_UNFREEZE)
        for i in range(start, total_blocks):
            for p in model.blocks[i].parameters():
                p.requires_grad = True
            # capture names for discriminative LR grouping heuristics
            unfrozen_block_names.append(f"blocks.{i}")
    elif hasattr(model, "transformer") and hasattr(model.transformer, "blocks"):
        total_blocks = len(model.transformer.blocks)
        print("Transformer blocks detected (model.transformer.blocks):", total_blocks)
        start = max(0, total_blocks - N_UNFREEZE)
        for i in range(start, total_blocks):
            for p in model.transformer.blocks[i].parameters():
                p.requires_grad = True
            unfrozen_block_names.append(f"transformer.blocks.{i}")
    else:
        # fallback: try to find parameters with 'block' in name and unfreeze last N found
        block_param_names = [n for n, _ in model.named_parameters() if "block" in n or "blocks" in n]
        total_blocks = len(block_param_names)
        print("Transformer block-like params found:", total_blocks)
        # best-effort: unfreeze last N param-groups containing 'block'
        last_names = block_param_names[-N_UNFREEZE:]
        for n, p in model.named_parameters():
            if any(x in n for x in last_names):
                p.requires_grad = True
                unfrozen_block_names.append(n)
    # If unfrozen set empty, fallback to making whole model trainable
    if not unfrozen_block_names:
        for p in model.parameters():
            p.requires_grad = True
        print("Fallback: made entire model trainable.")
except Exception as e:
    print("Unfreeze fallback — setting entire model trainable. Error:", e)
    for p in model.parameters():
        p.requires_grad = True

# Recreate optimizer and scheduler for finetuning (discriminative LRs)
param_groups_p2 = get_param_groups_phase2(model, unfrozen_blocks_names=unfrozen_block_names)
optimizer = AdamW(param_groups_p2, lr=HEAD_LR_PHASE2, weight_decay=WEIGHT_DECAY)
scheduler = make_epoch_scheduler(optimizer, total_epochs=PHASE2_EPOCHS + PHASE1_EPOCHS, warmup_epochs=WARMUP_EPOCHS)

# SAM wrapper for phase2
sam_obj = None
if USE_SAM:
    # collect the training parameters again
    trainable_params = [p for group in param_groups_p2 for p in group["params"]]
    base_optimizer_cls = AdamW
    sam_obj = SAM(params=trainable_params, base_optimizer=base_optimizer_cls, rho=0.05, adaptive=False, lr=HEAD_LR_PHASE2, weight_decay=WEIGHT_DECAY)
    optimizer_for_step = sam_obj.base_optimizer
else:
    optimizer_for_step = optimizer

# reset scaler for phase2
scaler = torch.cuda.amp.GradScaler(enabled=(torch.cuda.is_available() and not USE_SAM))

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer_for_step, scaler, device,
                                            epoch, total_epochs, amp_enabled=(not USE_SAM), accum_steps=ACCUM_STEPS,
                                            use_sam=USE_SAM, sam_obj=sam_obj)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    # step scheduler (epoch)
    scheduler.step()

    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Final evaluation ----------------
print("Loading best model from:", CHECKPOINT_PATH)
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# ---------------- Save plots ----------------
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(epochs_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, "inception_repl_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot to:", acc_path)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(epochs_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, "inception_repl_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot to:", loss_path)

cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, "inception_repl_confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()
print("Saved confusion matrix to:", cm_path)

plt.figure(figsize=(9,8))
plotted = False
for i in range(num_classes):
    if i in roc_auc_dict:
        plt.plot(fpr_dict[i], tpr_dict[i], label=f"{classes[i]} (AUC={roc_auc_dict[i]:.3f})"); plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title("ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, "inception_repl_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot to:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

# ---------------- Save final model ----------------
torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
