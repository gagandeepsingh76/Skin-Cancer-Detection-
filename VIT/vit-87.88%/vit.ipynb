```python
# MobileNetV3Large PyTorch training & evaluation (Colab-ready)
# Replaces your TensorFlow MobileNetV3 notebook with a fully-updated PyTorch pipeline.
# Requirements: torch, torchvision, timm (optional), scikit-learn, tqdm, matplotlib, seaborn
# Uncomment to install if needed:
# !pip install -q torch torchvision scikit-learn tqdm matplotlib seaborn

import os, random, zipfile, time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import transforms, datasets, models
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight
from tqdm import tqdm

# ---------------- Config (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/d/train"
VAL_DIR = "/content/dataset/d/val"
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_mobilenetv3_pytorch.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_mobilenetv3_pytorch.pt"

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2          # safe default for Colab
SEED = 42
PHASE1_EPOCHS = 8
PHASE2_EPOCHS = 18
N_UNFREEZE = 20         # number of later layers to unfreeze for fine-tuning (tune to dataset)
SHOW_GPU_MEM = True     # show GPU memory in postfix if CUDA available

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract dataset ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted.")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device & dataset checks ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found: {TRAIN_DIR} / {VAL_DIR}")

classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

# ---------------- Transforms ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Datasets & loaders ----------------
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Class weights ----------------
y_train = [label for _, label in train_dataset.samples]
if num_classes > 1 and len(y_train) > 0:
    cw_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(cw_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# ---------------- Model: MobileNetV3Large (torchvision) ----------------
model = models.mobilenet_v3_large(pretrained=True)
# replace classifier head to match num_classes
in_features = model.classifier[0].in_features if isinstance(model.classifier, nn.Sequential) else None
# Create a simple classifier similar to TF head
model.classifier = nn.Sequential(
    nn.Linear(model.classifier[0].in_features if hasattr(model.classifier[0],'in_features') else model.classifier[0].in_channels, 256),
    nn.Hardswish(),
    nn.Dropout(p=0.2),
    nn.Linear(256, num_classes)
)
model.to(device)
print("MobileNetV3Large model created with new classifier head.")

# ---------------- Freeze initial layers for phase 1 ----------------
# Freeze first layers, unfreeze later ones for fine-tuning
total_params = sum(p.numel() for p in model.parameters())
for p in model.parameters():
    p.requires_grad = False

# Make classifier params trainable
for name, p in model.named_parameters():
    if "classifier" in name:
        p.requires_grad = True

print("Classifier params set trainable.")

# ---------------- Loss / Optimizer / Scheduler ----------------
criterion = nn.CrossEntropyLoss(weight=class_weights) if class_weights is not None else nn.CrossEntropyLoss()
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-3, weight_decay=1e-5)  # head training lr
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

# ---------------- Utility functions ----------------
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list)
    y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1,1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

# ---------------- Single-line progress helpers ----------------
def train_one_epoch_single(model, loader, optimizer, device, epoch, total_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    for batch_idx, (imgs, labels) in pbar:
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        logits = model(imgs)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        b_acc = (preds == labels).float().mean().item()
        running_epoch_loss = running_loss / total
        running_epoch_acc = correct / total
        postfix = {
            "b_loss": f"{loss.item():.4f}",
            "b_acc": f"{b_acc:.3f}",
            "e_loss": f"{running_epoch_loss:.4f}",
            "e_acc": f"{running_epoch_acc:.3f}"
        }
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    return running_loss / len(loader.dataset), correct / total

def validate_single(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device)
            labels = labels.to(device)
            logits = model(imgs)
            loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return running_loss / len(loader.dataset), all_labels, all_preds, all_probs

# ---------------- Training loop ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0

total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS
print("Phase 1 (train head only) for", PHASE1_EPOCHS, "epochs.")
for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch_single(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate_single(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = val_loss
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Phase 2: unfreeze later layers for fine-tuning ----------------
print("Unfreezing later layers (last N) and starting fine-tuning.")
# Strategy: identify last N parameters groups (approx) — here we unfreeze part of features
# MobileNetV3 architecture: we can unfreeze last N layers by name scanning
all_named = list(model.named_children())
# Simple approach: unfreeze classifier (already trainable) and the last block of features
try:
    # If model.features exists and is Sequential, unfreeze last N modules
    if hasattr(model, "features") and isinstance(model.features, nn.Sequential):
        total_blocks = len(model.features)
        start_idx = max(0, total_blocks - N_UNFREEZE)
        for i in range(start_idx, total_blocks):
            for p in model.features[i].parameters():
                p.requires_grad = True
        print(f"Unfroze features[{start_idx}:] blocks.")
    else:
        # fallback: unfreeze entire model
        for p in model.parameters():
            p.requires_grad = True
        print("Fallback: made all params trainable.")
except Exception as e:
    print("Unfreeze fallback — setting whole model trainable. Error:", e)
    for p in model.parameters():
        p.requires_grad = True

# Recreate optimizer with lower LR for fine-tuning
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-5, weight_decay=1e-5)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch_single(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate_single(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Final evaluation ----------------
print("Loading best model from:", CHECKPOINT_PATH)
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate_single(model, val_loader, device, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# ---------------- Save plots ----------------
os.makedirs(GRAPH_DIR, exist_ok=True)
ep_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(ep_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(ep_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs")
plt.grid(True, linestyle='--', alpha=0.4); plt.legend(); plt.tight_layout()
acc_path = os.path.join(GRAPH_DIR, "mobilenetv3_accuracy.png"); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot:", acc_path)

plt.figure(figsize=(8,5))
plt.plot(ep_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(ep_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs")
plt.grid(True, linestyle='--', alpha=0.4); plt.legend(); plt.tight_layout()
loss_path = os.path.join(GRAPH_DIR, "mobilenetv3_loss.png"); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot:", loss_path)

cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, "mobilenetv3_confusion_matrix.png"); plt.savefig(cm_path, dpi=150); plt.show()
print("Saved confusion matrix:", cm_path)

plt.figure(figsize=(9,8))
plotted = False
for i in range(num_classes):
    if i in roc_auc_dict:
        plt.plot(fpr_dict[i], tpr_dict[i], label=f"{classes[i]} (AUC={roc_auc_dict[i]:.3f})"); plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title("ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, "mobilenetv3_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

# ---------------- Save final model ----------------
torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
