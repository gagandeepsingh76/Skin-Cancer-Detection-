```python
# Updated: ViT training with Gradual unfreezing + Discriminative (layer-wise) LR
# Keep dataset paths unchanged.
# Requirements: torch, torchvision, timm, scikit-learn, tqdm, matplotlib, seaborn

import os
import random
import zipfile
import time
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torchvision import transforms, datasets
import timm
from tqdm import tqdm
import re

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.utils.class_weight import compute_class_weight

# ---------------- Configuration (edit as needed) ----------------
DRIVE_MOUNT = True
ZIP_PATH = "/content/drive/MyDrive/d.zip"            # dataset zip on Drive (if used)
EXTRACT_PATH = "/content/dataset"
TRAIN_DIR = "/content/dataset/d/train"
VAL_DIR   = "/content/dataset/d/val"
GRAPH_DIR = "/content/drive/MyDrive/diabetic_eye_results"
CHECKPOINT_PATH = "/content/drive/MyDrive/best_vit_inception_replacement.pt"
FINAL_MODEL_PATH = "/content/drive/MyDrive/final_vit_inception_replacement.pt"

MODEL_NAME = "vit_base_patch16_224"   # timm ViT model
IMG_SIZE = 224                        # ViT standard
BATCH_SIZE = 32
NUM_WORKERS = 2                       # safe default for Colab
SEED = 42

# Progressive unfreezing / discriminative lr settings (NEW)
PHASE1_EPOCHS = 8        # phase1 (head-only) recommended 5-10
PHASE2_EPOCHS = 20       # phase2 (finetune) recommended 15-40
N_UNFREEZE = 8           # unfreeze last N transformer blocks in phase2 (6-12 suggested)
SHOW_GPU_MEM = True      # show GPU memory in postfix if CUDA available

# Layer-wise LR hyperparameters (NEW)
HEAD_LR = 3e-4           # head training LR (phase1) - set 1e-4 to 3e-4
FINETUNE_LR = 5e-5       # base LR for finetuning (phase2) - set in 1e-6 .. 5e-5
LAYER_DECAY = 0.8        # layer-wise decay factor (0.7-0.9 recommended)
WEIGHT_DECAY = 1e-2

# ---------------- Reproducibility ----------------
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ---------------- Mount Drive & extract dataset (Colab) ----------------
if DRIVE_MOUNT:
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=False)
    except Exception:
        pass

if os.path.exists(ZIP_PATH) and not os.path.exists(EXTRACT_PATH):
    with zipfile.ZipFile(ZIP_PATH, 'r') as z:
        z.extractall(EXTRACT_PATH)
        print("Dataset extracted to", EXTRACT_PATH)
else:
    print("Zip not found or dataset already extracted (or already extracted).")

os.makedirs(GRAPH_DIR, exist_ok=True)
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

# ---------------- Device & dataset checks ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if not os.path.isdir(TRAIN_DIR) or not os.path.isdir(VAL_DIR):
    raise FileNotFoundError(f"Train/Val directories not found. Expected: {TRAIN_DIR} and {VAL_DIR}")

# ---------------- Class discovery ----------------
classes = sorted([d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))])
num_classes = len(classes)
print(f"Detected {num_classes} classes: {classes}")

# ---------------- Transforms ----------------
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.12, contrast=0.12, saturation=0.12, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# ---------------- Datasets & DataLoaders ----------------
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                                           num_workers=NUM_WORKERS, pin_memory=True)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset))

# ---------------- Class weights ----------------
y_train = [label for _, label in train_dataset.samples]
if num_classes > 1 and len(y_train) > 0:
    class_weights_np = compute_class_weight(class_weight='balanced', classes=np.arange(num_classes), y=y_train)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print("Class weights:", class_weights.cpu().numpy())
else:
    class_weights = None

# ---------------- Model (timm ViT) ----------------
model = timm.create_model(MODEL_NAME, pretrained=True, num_classes=num_classes)
model.to(device)
print("Model created:", MODEL_NAME)

# ---------------- Replace / ensure head is trainable in phase1 ----------------
# Freeze encoder backbone parameters
for name, p in model.named_parameters():
    p.requires_grad = False

# Unfreeze classifier head parameters (common names used by timm)
head_param_count = 0
for name, p in model.named_parameters():
    if ("head" in name) or ("classifier" in name) or name.endswith("fc") or "norm" in name and "head" in name:
        p.requires_grad = True
        head_param_count += 1

# fallback: if none found make last linear layers trainable
if head_param_count == 0:
    lst = list(model.named_parameters())[-40:]
    for name, p in lst:
        p.requires_grad = True
        head_param_count += 1

print("Approx. trainable head parameter groups:", head_param_count)

# ---------------- Loss ----------------
criterion = nn.CrossEntropyLoss(weight=class_weights) if class_weights is not None else nn.CrossEntropyLoss()

# ---------------- Utility: compute layer id for ViT params ----------------
def get_layer_id_for_vit(param_name):
    """
    Return layer id for ViT-like models.
    - patch_embed / pos_embed -> layer 0
    - blocks.i.* -> layer i+1
    - head/classifier -> max_layer + 2 (head group)
    """
    m = re.search(r'blocks\.(\d+)\.', param_name)
    if m:
        return int(m.group(1)) + 1
    # positional embedding / patch embedding
    if param_name.startswith('patch_embed') or 'pos_embed' in param_name:
        return 0
    # default head / classifier group
    if any(k in param_name for k in ['head', 'classifier', 'fc', 'norm']):
        return 1000
    # fallback
    return 0

def get_parameter_groups(model, base_lr, weight_decay, layer_decay):
    """
    Build param groups for discriminative lr (layer-wise decay).
    Returns list of dicts accepted by optimizer.
    """
    # compute layer ids for all trainable params
    param_layer_map = {}
    layer_ids = []
    for n, p in model.named_parameters():
        if not p.requires_grad:
            continue
        lid = get_layer_id_for_vit(n)
        param_layer_map.setdefault(lid, []).append((n, p))
        layer_ids.append(lid)
    if len(layer_ids) == 0:
        return [{"params": [p for n,p in model.named_parameters() if p.requires_grad], "lr": base_lr, "weight_decay": weight_decay}]
    max_id = max(layer_ids)
    groups = []
    # for deterministic ordering, sort layer ids ascending (0 -> ... -> head)
    for lid in sorted(param_layer_map.keys()):
        # scale: deeper layers (larger lid) get larger LR; compute exponent relative to max_id
        # scale = layer_decay ** (max_id - lid)
        scale = layer_decay ** (max_id - lid)
        lr = base_lr * scale
        group = {"params": [p for _, p in param_layer_map[lid]], "lr": lr, "weight_decay": weight_decay}
        groups.append(group)
    return groups

# ---------------- Training & validation functions (unchanged except optimizer uses param groups) ----------------
def human_gpu_mem():
    if not torch.cuda.is_available() or not SHOW_GPU_MEM:
        return ""
    return f"{torch.cuda.memory_reserved(device)/1024**3:.2f}GB"

def compute_metrics_all(y_true_list, y_pred_list, y_prob_array, class_names):
    y_true = np.array(y_true_list)
    y_pred = np.array(y_pred_list)
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    if y_true_bin.ndim == 1:
        y_true_bin = y_true_bin.reshape(-1, 1)
    if y_true_bin.shape[1] == 1:
        y_true_bin = np.hstack([1 - y_true_bin, y_true_bin])
    fpr, tpr, roc_auc = {}, {}, {}
    aucs = []
    missing = []
    for i in range(n_classes):
        if np.sum(y_true_bin[:, i]) == 0:
            missing.append(class_names[i])
            continue
        try:
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_array[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            aucs.append(roc_auc[i])
        except Exception:
            missing.append(class_names[i])
    macro_auc = np.mean(aucs) if len(aucs) > 0 else 0.0
    return acc, report, roc_auc, fpr, tpr, macro_auc, missing

def train_one_epoch(model, loader, optimizer, device, epoch, total_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Train {epoch}/{total_epochs}", ncols=120, leave=False)
    for batch_idx, (imgs, labels) in pbar:
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        logits = model(imgs)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
        b_acc = (preds == labels).float().mean().item()
        running_epoch_loss = running_loss / total
        running_epoch_acc = correct / total
        postfix = {
            "b_loss": f"{loss.item():.4f}",
            "b_acc": f"{b_acc:.3f}",
            "e_loss": f"{running_epoch_loss:.4f}",
            "e_acc": f"{running_epoch_acc:.3f}"
        }
        if torch.cuda.is_available() and SHOW_GPU_MEM:
            postfix["GPU"] = human_gpu_mem()
        pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    return running_loss / len(loader.dataset), correct / total if total>0 else 0.0

def validate(model, loader, device, epoch=None):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    all_probs = []
    pbar = tqdm(enumerate(loader, start=1), total=len(loader),
                desc=f"Val {epoch}" if epoch else "Val", ncols=120, leave=False)
    with torch.no_grad():
        for batch_idx, (imgs, labels) in pbar:
            imgs = imgs.to(device)
            labels = labels.to(device)
            logits = model(imgs)
            loss = criterion(logits, labels)
            running_loss += loss.item() * imgs.size(0)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            preds = np.argmax(probs, axis=1)
            all_probs.append(probs)
            all_preds.extend(list(preds))
            all_labels.extend(list(labels.cpu().numpy()))
            batch_acc = (preds == labels.cpu().numpy()).mean()
            running_epoch_loss = running_loss / (len(all_labels) if len(all_labels)>0 else 1)
            running_epoch_acc = accuracy_score(all_labels, all_preds) if len(all_labels)>0 else 0.0
            postfix = {
                "b_loss": f"{loss.item():.4f}",
                "b_acc": f"{batch_acc:.3f}",
                "e_loss": f"{running_epoch_loss:.4f}",
                "e_acc": f"{running_epoch_acc:.3f}"
            }
            if torch.cuda.is_available() and SHOW_GPU_MEM:
                postfix["GPU"] = human_gpu_mem()
            pbar.set_postfix(postfix, refresh=False)
    pbar.close()
    val_loss = running_loss / len(loader.dataset) if len(loader.dataset) > 0 else 0.0
    all_probs = np.vstack(all_probs) if all_probs else np.zeros((0, num_classes))
    return val_loss, all_labels, all_preds, all_probs

# ---------------- Training loop (Phase1 head-only, Phase2 progressive unfreeze + layer-wise LR) ----------------
history = {"train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_auc_macro":[]}
best_val_loss = float('inf')
best_auc_macro = 0.0
total_epochs = PHASE1_EPOCHS + PHASE2_EPOCHS

# --- Phase 1: head-only training ---
print("Starting Phase 1 (head-only) for", PHASE1_EPOCHS, "epochs.")
# Create optimizer for head parameters only
head_params = list(filter(lambda p: p.requires_grad, model.parameters()))
if len(head_params) == 0:
    raise RuntimeError("No head parameters set to requires_grad=True for phase1. Check model head naming.")
optimizer = AdamW(head_params, lr=HEAD_LR, weight_decay=WEIGHT_DECAY)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

for epoch in range(1, PHASE1_EPOCHS + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{PHASE1_EPOCHS} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = val_loss
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# --- Phase 2: progressive unfreeze + discriminative lr ---
print("Starting Phase 2: unfreeze last", N_UNFREEZE, "transformer blocks and apply layer-wise LR.")
# Unfreeze last N_UNFREEZE blocks if available
unfrozen_count = 0
try:
    if hasattr(model, "blocks"):
        total_blocks = len(model.blocks)
        print("Transformer blocks detected:", total_blocks)
        start_block = max(0, total_blocks - N_UNFREEZE)
        for i in range(start_block, total_blocks):
            for p in model.blocks[i].parameters():
                p.requires_grad = True
                unfrozen_count += 1
    elif hasattr(model, "transformer") and hasattr(model.transformer, "blocks"):
        total_blocks = len(model.transformer.blocks)
        print("Transformer blocks detected (transformer.blocks):", total_blocks)
        start_block = max(0, total_blocks - N_UNFREEZE)
        for i in range(start_block, total_blocks):
            for p in model.transformer.blocks[i].parameters():
                p.requires_grad = True
                unfrozen_count += 1
    else:
        # fallback: unfreeze entire model (if no blocks found)
        for p in model.parameters():
            p.requires_grad = True
        print("No transformer blocks found; made whole model trainable as fallback.")
except Exception as e:
    print("Unfreeze fallback — setting entire model trainable. Error:", e)
    for p in model.parameters():
        p.requires_grad = True

print(f"Unfrozen parameters (approx): {unfrozen_count}")

# Recreate optimizer with layer-wise parameter groups
# If ViT-like, get layer-wise groups; otherwise fall back to single group
trainable_params = [p for n, p in model.named_parameters() if p.requires_grad]
if any("blocks" in n for n, _ in model.named_parameters()):
    param_groups = get_parameter_groups(model, base_lr=FINETUNE_LR, weight_decay=WEIGHT_DECAY, layer_decay=LAYER_DECAY)
else:
    # single group
    param_groups = [{"params": trainable_params, "lr": FINETUNE_LR, "weight_decay": WEIGHT_DECAY}]

optimizer = AdamW(param_groups, lr=FINETUNE_LR, weight_decay=WEIGHT_DECAY)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=2)

print("Starting Phase 2 (fine-tuning) for", PHASE2_EPOCHS, "epochs.")
for epoch in range(PHASE1_EPOCHS + 1, total_epochs + 1):
    t0 = time.time()
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device, epoch, total_epochs)
    val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch)
    _, _, _, _, _, val_auc_macro, _ = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
    scheduler.step(val_loss)
    history["train_loss"].append(train_loss); history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(accuracy_score(y_true_val, y_pred_val))
    history["val_auc_macro"].append(val_auc_macro)
    print(f"Epoch {epoch}/{total_epochs} summary: train_loss={train_loss:.4f} train_acc={train_acc:.4f} val_loss={val_loss:.4f} val_acc={history['val_acc'][-1]:.4f} val_auc={val_auc_macro:.4f}  time={time.time()-t0:.1f}s")
    if val_loss < best_val_loss or val_auc_macro > best_auc_macro:
        best_val_loss = min(best_val_loss, val_loss)
        best_auc_macro = max(best_auc_macro, val_auc_macro)
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print("Saved best model to:", CHECKPOINT_PATH)

# ---------------- Final evaluation ----------------
print("Loading best model from:", CHECKPOINT_PATH)
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=device))
model.to(device)
model.eval()

val_loss, y_true_val, y_pred_val, y_prob_val = validate(model, val_loader, device, epoch="final")
acc, report, roc_auc_dict, fpr_dict, tpr_dict, macro_auc, missing = compute_metrics_all(y_true_val, y_pred_val, y_prob_val, classes)
print("\nFinal Validation Accuracy: {:.2f}%".format(acc * 100))
print("\nClassification Report:\n", report)
print("Macro AUC (mean of available class AUCs):", macro_auc)
if missing:
    print("Skipped ROC for classes (no positive samples):", missing)

# ---------------- Save plots ----------------
os.makedirs(GRAPH_DIR, exist_ok=True)
epochs_range = range(1, len(history["train_loss"]) + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_acc"], marker='o', label='Train Acc')
plt.plot(epochs_range, history["val_acc"], marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
acc_path = os.path.join(GRAPH_DIR, "inception_repl_accuracy.png")
plt.tight_layout(); plt.savefig(acc_path, dpi=150); plt.show()
print("Saved accuracy plot to:", acc_path)

plt.figure(figsize=(8,5))
plt.plot(epochs_range, history["train_loss"], marker='o', label='Train Loss')
plt.plot(epochs_range, history["val_loss"], marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss over Epochs"); plt.grid(True, linestyle='--', alpha=0.4); plt.legend()
loss_path = os.path.join(GRAPH_DIR, "inception_repl_loss.png")
plt.tight_layout(); plt.savefig(loss_path, dpi=150); plt.show()
print("Saved loss plot to:", loss_path)

cm = confusion_matrix(y_true_val, y_pred_val, labels=list(range(num_classes)))
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title("Confusion Matrix"); plt.tight_layout()
cm_path = os.path.join(GRAPH_DIR, "inception_repl_confusion_matrix.png")
plt.savefig(cm_path, dpi=150); plt.show()
print("Saved confusion matrix to:", cm_path)

plt.figure(figsize=(9,8))
plotted = False
for i in range(num_classes):
    if i in roc_auc_dict:
        plt.plot(fpr_dict[i], tpr_dict[i], label=f"{classes[i]} (AUC={roc_auc_dict[i]:.3f})"); plotted = True
plt.plot([0,1],[0,1], linestyle='--', label='Chance')
if plotted:
    plt.title("ROC Curves per class"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(loc='lower right')
    roc_path = os.path.join(GRAPH_DIR, "inception_repl_roc.png"); plt.tight_layout(); plt.savefig(roc_path, dpi=150); plt.show()
    print("Saved ROC plot to:", roc_path)
else:
    print("No ROC curves plotted (no positive samples for some classes).")

# ---------------- Save final model ----------------
torch.save(model.state_dict(), FINAL_MODEL_PATH)
print("Saved final model state dict to:", FINAL_MODEL_PATH)
```
